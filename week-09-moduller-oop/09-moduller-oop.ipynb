# **9.6 self KavramÄ±**

`self`, sÄ±nÄ±f iÃ§inde **mevcut nesneyi temsil eder**. MetotlarÄ±n ilk parametresi her zaman `self` olmalÄ±dÄ±r.

- `self.nitelik` - Nesnenin niteliÄŸine eriÅŸim
- `self.metot()` - Nesnenin metodunu Ã§aÄŸÄ±rma

````python
# self kullanÄ±mÄ± - metotlarda
# self, metodun Ã§aÄŸrÄ±ldÄ±ÄŸÄ± nesneyi temsil eder

class Dikdortgen:
    """DikdÃ¶rtgen hesaplamalarÄ±."""
    
    def __init__(self, uzunluk, genislik):
        """BoyutlarÄ± ata."""
        self.uzunluk = uzunluk  # self ile nesneye baÄŸla
        self.genislik = genislik
    
    def alan(self):
        """Alan hesapla."""
        # self.uzunluk ve self.genislik: bu nesnenin deÄŸerleri
        return self.uzunluk * self.genislik
    
    def cevre(self):
        """Ã‡evre hesapla."""
        return 2 * (self.uzunluk + self.genislik)
    
    def bilgi(self):
        """Bilgileri gÃ¶ster."""
        # DiÄŸer metotlarÄ± self ile Ã§aÄŸÄ±r
        print(f"Boyutlar: {self.uzunluk} x {self.genislik}")
        print(f"Alan: {self.alan()} mÂ²")
        print(f"Ã‡evre: {self.cevre()} m")

# Nesne oluÅŸtur ve metotlarÄ± Ã§aÄŸÄ±r
d1 = Dikdortgen(5, 3)
d1.bilgi()

print()

d2 = Dikdortgen(10, 4)
d2.bilgi()
````

````python
# SÄ±nÄ±f niteliÄŸi vs Ã–rnek niteliÄŸi
# SÄ±nÄ±f niteliÄŸi: tÃ¼m Ã¶rnekler iÃ§in ortak
# Ã–rnek niteliÄŸi: her nesne iÃ§in farklÄ±

class Ogrenci:
    """Ã–ÄŸrenci sÄ±nÄ±fÄ±."""
    
    okul = "MYO"       # SÄ±nÄ±f niteliÄŸi (ortak)
    toplam_ogrenci = 0  # SÄ±nÄ±f niteliÄŸi (ortak sayaÃ§)
    
    def __init__(self, ad, numara):
        self.ad = ad           # Ã–rnek niteliÄŸi (farklÄ±)
        self.numara = numara   # Ã–rnek niteliÄŸi (farklÄ±)
        Ogrenci.toplam_ogrenci += 1  # SÄ±nÄ±f niteliÄŸini gÃ¼ncelle
    
    def bilgi(self):
        print(f"{self.ad} ({self.numara}) - {Ogrenci.okul}")

# Ã–ÄŸrenci oluÅŸtur
o1 = Ogrenci("Ali", 101)
o2 = Ogrenci("Veli", 102)
o3 = Ogrenci("AyÅŸe", 103)

print(f"Toplam Ã¶ÄŸrenci: {Ogrenci.toplam_ogrenci}")
print()

o1.bilgi()
o2.bilgi()
o3.bilgi()
````

# **9.7 KalÄ±tÄ±m (Inheritance)**

**KalÄ±tÄ±m**, bir sÄ±nÄ±fÄ±n baÅŸka bir sÄ±nÄ±fÄ±n Ã¶zelliklerini **miras almasÄ±**dÄ±r.

- **Taban sÄ±nÄ±f (Parent)**: Miras veren sÄ±nÄ±f
- **Alt sÄ±nÄ±f (Child)**: Miras alan sÄ±nÄ±f
- `super()`: Taban sÄ±nÄ±fa eriÅŸim saÄŸlar

```python
class AltSinif(TabanSinif):
    def __init__(self, ...):
        super().__init__(...)  # Taban sÄ±nÄ±fÄ±n __init__'ini Ã§aÄŸÄ±r
```

````python
# KalÄ±tÄ±m Ã¶rneÄŸi
# Alt sÄ±nÄ±f, taban sÄ±nÄ±fÄ±n Ã¶zelliklerini miras alÄ±r

# Taban sÄ±nÄ±f (parent class)
class Hayvan:
    """TÃ¼m hayvanlar iÃ§in temel sÄ±nÄ±f."""
    
    def __init__(self, isim):
        self.isim = isim
    
    def ses_cikar(self):
        print("Hayvan ses Ã§Ä±karÄ±yor...")
    
    def bilgi(self):
        print(f"Hayvan: {self.isim}")

# Alt sÄ±nÄ±f (child class)
class Kopek(Hayvan):  # Hayvan'dan miras al
    """KÃ¶pek sÄ±nÄ±fÄ± - Hayvan'dan tÃ¼retildi."""
    
    def __init__(self, isim, cins):
        super().__init__(isim)  # Taban sÄ±nÄ±fÄ±n __init__'ini Ã§aÄŸÄ±r
        self.cins = cins        # Ek nitelik
    
    # Metot override (geÃ§ersiz kÄ±lma)
    def ses_cikar(self):
        print(f"{self.isim} havlÄ±yor: Hav hav!")
    
    # Yeni metot
    def kostur(self):
        print(f"{self.isim} koÅŸuyor!")

# Alt sÄ±nÄ±f
class Kedi(Hayvan):
    """Kedi sÄ±nÄ±fÄ±."""
    
    def ses_cikar(self):
        print(f"{self.isim} miyavlÄ±yor: Miyav!")

# Nesneler oluÅŸtur
kopek = Kopek("KarabaÅŸ", "Golden")
kedi = Kedi("Pamuk")

# MetotlarÄ± Ã§aÄŸÄ±r
kopek.bilgi()      # Miras alÄ±nan metot
kopek.ses_cikar()  # Override edilmiÅŸ metot
kopek.kostur()     # Yeni metot

print()

kedi.bilgi()
kedi.ses_cikar()
````

# **9.8 Pratik Ã–rnekler**

````python
# Pratik Ã–rnek 1: Banka HesabÄ± SÄ±nÄ±fÄ±
# Para yatÄ±rma, Ã§ekme ve bakiye kontrolÃ¼

class BankaHesabi:
    """Banka hesabÄ± yÃ¶netimi."""
    
    # SÄ±nÄ±f niteliÄŸi: tÃ¼m hesaplar iÃ§in ortak faiz oranÄ±
    faiz_orani = 0.05
    
    def __init__(self, hesap_no, sahip, bakiye=0):
        """Hesap oluÅŸtur."""
        self.hesap_no = hesap_no
        self.sahip = sahip
        self._bakiye = bakiye  # _ ile "Ã¶zel" olduÄŸunu belirt
    
    def para_yatir(self, miktar):
        """Para yatÄ±r."""
        if miktar > 0:
            self._bakiye += miktar
            print(f"{miktar} TL yatÄ±rÄ±ldÄ±. Yeni bakiye: {self._bakiye} TL")
        else:
            print("GeÃ§ersiz miktar!")
    
    def para_cek(self, miktar):
        """Para Ã§ek."""
        if miktar > self._bakiye:
            print("Yetersiz bakiye!")
        elif miktar <= 0:
            print("GeÃ§ersiz miktar!")
        else:
            self._bakiye -= miktar
            print(f"{miktar} TL Ã§ekildi. Kalan bakiye: {self._bakiye} TL")
    
    def bakiye_sorgula(self):
        """Bakiyeyi gÃ¶ster."""
        return self._bakiye
    
    def faiz_ekle(self):
        """YÄ±llÄ±k faizi ekle."""
        faiz = self._bakiye * BankaHesabi.faiz_orani
        self._bakiye += faiz
        print(f"{faiz:.2f} TL faiz eklendi. Yeni bakiye: {self._bakiye:.2f} TL")

# Test
hesap = BankaHesabi("TR001", "ZÃ¼ber DoÄŸan", 1000)
hesap.para_yatir(500)
hesap.para_cek(200)
hesap.faiz_ekle()
print(f"\nGÃ¼ncel bakiye: {hesap.bakiye_sorgula():.2f} TL")
````

````python
# Pratik Ã–rnek 2: MÃ¼ÅŸteri YÃ¶netim Sistemi
# SÄ±nÄ±f metodu ve sÄ±nÄ±f deÄŸiÅŸkeni kullanÄ±mÄ±

class Musteri:
    """MÃ¼ÅŸteri yÃ¶netim sistemi."""
    
    # SÄ±nÄ±f niteliÄŸi: tÃ¼m mÃ¼ÅŸterileri sakla
    tum_musteriler = []
    
    def __init__(self, isim, email):
        self.isim = isim
        self.email = email
        self.satin_almalar = []
        Musteri.tum_musteriler.append(self)  # Listeye ekle
    
    def satin_al(self, urun, fiyat):
        """ÃœrÃ¼n satÄ±n al."""
        self.satin_almalar.append({"urun": urun, "fiyat": fiyat})
        print(f"{self.isim}: {urun} satÄ±n aldÄ± ({fiyat} TL)")
    
    def toplam_harcama(self):
        """Toplam harcamayÄ± hesapla."""
        return sum(item["fiyat"] for item in self.satin_almalar)
    
    @classmethod  # SÄ±nÄ±f metodu
    def musteri_listele(cls):
        """TÃ¼m mÃ¼ÅŸterileri listele."""
        print("\n=== MÃ¼ÅŸteri Listesi ===")
        for m in cls.tum_musteriler:
            print(f"  - {m.isim} ({m.email})")
    
    @classmethod
    def toplam_musteri(cls):
        """Toplam mÃ¼ÅŸteri sayÄ±sÄ±nÄ± dÃ¶ndÃ¼r."""
        return len(cls.tum_musteriler)

# Test
m1 = Musteri("Ali YÄ±lmaz", "ali@mail.com")
m2 = Musteri("AyÅŸe Kaya", "ayse@mail.com")

m1.satin_al("Laptop", 15000)
m1.satin_al("Mouse", 200)
m2.satin_al("Telefon", 8000)

print(f"\n{m1.isim} toplam harcama: {m1.toplam_harcama()} TL")

Musteri.musteri_listele()
print(f"\nToplam mÃ¼ÅŸteri: {Musteri.toplam_musteri()}")
````

````python
# ============================================================================
# ðŸŸ¨ CEVAP ANAHTARI â€“ ModÃ¼ller ve Nesne TabanlÄ± Programlama
# ============================================================================
# YukarÄ±daki gÃ¶revlerin tamamlanmÄ±ÅŸ Ã§Ã¶zÃ¼mleri aÅŸaÄŸÄ±da verilmiÅŸtir.
# ============================================================================




# ============================================================================
# ðŸŸ¨ CEVAP 1: import ile modÃ¼l kullanÄ±mÄ±
# ============================================================================
import os
print(f"Ä°ÅŸletim sistemi: {os.name}")
print(f"Ã‡alÄ±ÅŸma dizini: {os.getcwd()}")

print(f"\n{'-'*82}\n")




# ============================================================================
# ðŸŸ¨ CEVAP 2: from ... import ile belirli Ã¶ÄŸe alma
# ============================================================================
from math import sqrt, pi, pow
print(f"KarekÃ¶k 25: {sqrt(25)}")    # 5.0
print(f"Pi: {pi}")                   # 3.14159...
print(f"2^8: {pow(2, 8)}")          # 256.0

print(f"\n{'-'*82}\n")




# ============================================================================
# ðŸŸ¨ CEVAP 3: Takma ad (alias) ile import
# ============================================================================
import datetime as dt
bugun = dt.date.today()
simdi = dt.datetime.now()
print(f"BugÃ¼n: {bugun}")
print(f"Åžimdi: {simdi}")

print(f"\n{'-'*82}\n")




# ============================================================================
# ðŸŸ¨ CEVAP 4: random modÃ¼lÃ¼
# ============================================================================
import random
print(f"Rastgele sayÄ±: {random.randint(1, 50)}")
renkler = ["kÄ±rmÄ±zÄ±", "mavi", "yeÅŸil"]
print(f"Rastgele renk: {random.choice(renkler)}")
sayilar = [1, 2, 3, 4, 5]
random.shuffle(sayilar)
print(f"KarÄ±ÅŸtÄ±rÄ±lmÄ±ÅŸ: {sayilar}")

print(f"\n{'-'*82}\n")




# ============================================================================
# ðŸŸ¨ CEVAP 5: dir() ile modÃ¼l iÃ§eriÄŸini gÃ¶rme
# ============================================================================
import math
icerik = [item for item in dir(math) if not item.startswith('_')][:10]
print(icerik)

print(f"\n{'-'*82}\n")




# ============================================================================
# ðŸŸ¨ CEVAP 6: Basit sÄ±nÄ±f tanÄ±mlama â€“ SÄ±nÄ±f nitelikleri
# ============================================================================
class Okul:
    ad = "MYO"
    sehir = "Ä°stanbul"
print(f"Okul: {Okul.ad}")
print(f"Åžehir: {Okul.sehir}")

print(f"\n{'-'*82}\n")




# ============================================================================
# ðŸŸ¨ CEVAP 7: __init__ metodu ve nesne oluÅŸturma
# ============================================================================
class Araba:
    def __init__(self, marka, model, yil):
        self.marka = marka
        self.model = model
        self.yil = yil
a1 = Araba("Toyota", "Corolla", 2023)
a2 = Araba("Honda", "Civic", 2022)
print(f"{a1.marka} {a1.model} ({a1.yil})")
print(f"{a2.marka} {a2.model} ({a2.yil})")

print(f"\n{'-'*82}\n")




# ============================================================================
# ðŸŸ¨ CEVAP 8: self ile metot tanÄ±mlama
# ============================================================================
class Kare:
    def __init__(self, kenar):
        self.kenar = kenar
    def alan(self):
        return self.kenar ** 2
    def cevre(self):
        return 4 * self.kenar
    def bilgi(self):
        print(f"Kenar: {self.kenar}")
        print(f"Alan: {self.alan()}")
        print(f"Ã‡evre: {self.cevre()}")
k = Kare(5)
k.bilgi()

print(f"\n{'-'*82}\n")




# ============================================================================
# ðŸŸ¨ CEVAP 9: SÄ±nÄ±f niteliÄŸi vs Ã–rnek niteliÄŸi
# ============================================================================
class Urun:
    magaza = "TeknoMarket"
    toplam_urun = 0
    def __init__(self, ad, fiyat):
        self.ad = ad
        self.fiyat = fiyat
        Urun.toplam_urun += 1
u1 = Urun("Laptop", 25000)
u2 = Urun("Telefon", 15000)
u3 = Urun("Tablet", 10000)
print(f"MaÄŸaza: {Urun.magaza}")
print(f"Toplam Ã¼rÃ¼n: {Urun.toplam_urun}")  # 3

print(f"\n{'-'*82}\n")




# ============================================================================
# ðŸŸ¨ CEVAP 10: KalÄ±tÄ±m (Inheritance)
# ============================================================================
class Sekil:
    def __init__(self, renk):
        self.renk = renk
    def bilgi(self):
        print(f"Renk: {self.renk}")

class Daire(Sekil):
    def __init__(self, renk, yaricap):
        super().__init__(renk)
        self.yaricap = yaricap
    def alan(self):
        return 3.14 * self.yaricap ** 2

class KareS(Sekil):
    def __init__(self, renk, kenar):
        super().__init__(renk)
        self.kenar = kenar
    def alan(self):
        return self.kenar ** 2

d = Daire("kÄ±rmÄ±zÄ±", 5)
d.bilgi()
print(f"Daire alanÄ±: {d.alan():.2f}")
k = KareS("mavi", 4)
k.bilgi()
print(f"Kare alanÄ±: {k.alan()}")

print(f"\n{'-'*82}\n")




# ============================================================================
# ðŸŸ¨ CEVAP 11: Pratik Ã–rnek â€“ Banka hesabÄ±
# ============================================================================
class Hesap:
    def __init__(self, sahip, bakiye=0):
        self.sahip = sahip
        self.bakiye = bakiye
    def yatir(self, miktar):
        self.bakiye += miktar
        print(f"{miktar} TL yatÄ±rÄ±ldÄ±.")
    def cek(self, miktar):
        if miktar > self.bakiye:
            print("Yetersiz bakiye!")
        else:
            self.bakiye -= miktar
            print(f"{miktar} TL Ã§ekildi.")
    def bakiye_goster(self):
        print(f"{self.sahip}: {self.bakiye} TL")
h = Hesap("Ali", 1000)
h.yatir(500)
h.cek(200)
h.bakiye_goster()  # Ali: 1300 TL
````
