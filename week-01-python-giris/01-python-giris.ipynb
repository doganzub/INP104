# **1.1 print() Fonksiyonu**

`print()` fonksiyonu, Python'da ekrana çıktı vermemizi sağlayan temel fonksiyondur. Etkileşimli kabukta doğrudan yazdığımız ifadeler ekranda görünse de, Python programlarımızı bir dosyaya kaydedip çalıştırdığımızda, başında `print()` olmayan ifadeler çıktıda görünmez.

Bu fonksiyon, tıpkı `type()` ve `len()` gibi Python'ın gömülü (built-in) fonksiyonlarından biridir. Fonksiyonların parantezleri içinde belirtilen öğelere **parametre** adı verilir.

---

### print() Fonksiyonunun Temel Kullanımı

`print()` fonksiyonu bir veya birden fazla parametre alabilir. Birden fazla parametre verildiğinde, bunlar varsayılan olarak aralarına birer boşluk karakteri yerleştirilerek birleştirilir.

```python
# print() fonksiyonu ile ekrana metin yazdırma
# Fonksiyon, parantez içindeki karakter dizisini ekrana basar
print("Merhaba Dünya!")  # En basit kullanım: tek parametre ile
```

```python
# print() fonksiyonu birden fazla parametre alabilir
# Parametreler virgül ile ayrılır ve aralarına otomatik olarak boşluk eklenir
print("Python", "Programlama", "Dersi")  # Üç ayrı parametre, boşlukla birleştirilir
```

```python
# Değişken kullanarak print() ile yazdırma
# Önce değer bir değişkene atanır, sonra değişken print() ile yazdırılır
dil = "Python programlama dili"  # Karakter dizisini değişkene atama
print(dil)  # Değişkenin değerini ekrana yazdırma
```

# **1.2 Tırnak Kullanımı**

Python'da karakter dizisi (string) tanımlamak için üç farklı tırnak seçeneği sunulur:

1. **Tek tırnak** (`' '`)
2. **Çift tırnak** (`" "`)
3. **Üç tırnak** (`""" """` veya `''' '''`)

Bu üç tırnak tipinin çıktıları arasında hiçbir fark yoktur. Peki neden üç farklı tırnak çeşidi vardır?

---

### Tırnak Seçiminin Önemi

Karakter dizisi içinde kesme işareti veya tırnak işareti geçtiğinde, Python'ın karakter dizisini başlatan ve bitiren tırnakları ayırt edebilmesi için farklı tırnak tipleri kullanılır. Python, karakter dizisini hangi tırnakla başlatmışsak, o tırnakla bitirmemizi bekler.

```python
# Tek tırnak kullanımı
# Basit karakter dizileri için tek tırnak yeterlidir
print('Merhaba Python')  # Tek tırnak ile karakter dizisi tanımlama
```

```python
# Çift tırnak içinde tek tırnak (kesme işareti) kullanımı
# Karakter dizisi içinde kesme işareti varsa, dışta çift tırnak kullanılır
# Böylece Python hangi tırnağın karakter dizisini başlatıp bitirdiğini ayırt edebilir
print("İstanbul'un 5 günlük hava durumu tahmini")  # Kesme işareti problem yaratmaz
```

```python
# Tek tırnak içinde çift tırnak kullanımı
# Karakter dizisi içinde çift tırnak varsa, dışta tek tırnak kullanılır
# Bu sayede iç içe tırnak işaretleri birbirine karışmaz
print('Python programlama dilinin adı "piton" yılanından gelmez')  # İç tırnak sorunsuz
```

```python
# Üç tırnak kullanımı - Çok satırlı metin için idealdir
# Üç tırnak, Enter tuşuna basıldığında hata vermez ve satır atlamaya izin verir
# Genellikle birden fazla satıra yayılmış metinlerde tercih edilir
print("""Bu bir
çok satırlı
metin örneğidir.""")  # Her satır olduğu gibi yazdırılır
```

```python
# Üç tırnak ile karmaşık metin örneği
# Hem tek tırnak hem çift tırnak içeren metinler için üç tırnak kullanılabilir
print("""Python programlama dili Guido Van Rossum
adlı Hollandalı bir programcı tarafından 90'lı
yılların başında geliştirilmeye başlanmıştır. Çoğu
insan, isminin "Python" olmasına bakarak, bu programlama
dilinin, adını piton yılanından aldığını düşünür.
Ancak zannedildiğinin aksine bu programlama dilinin
adı piton yılanından gelmez.""")  # Hem kesme hem çift tırnak kullanılabilir
```

# **1.3 Veri Tipleri**

Python'da her değerin bir **veri tipi** vardır. Programlama sürecinde verilerle çalışırken, bu verilerin hangi tipte olduğunu bilmek son derece önemlidir. Çünkü farklı veri tipleri üzerinde farklı işlemler yapılabilir.

---

### Temel Veri Tipleri

Python'da en sık kullanılan temel veri tipleri şunlardır:

| Veri Tipi | Python Kodu | Açıklama |
|-----------|-------------|----------|
| **Karakter Dizisi** | `str` | Metin verileri (harfler, kelimeler, cümleler) |
| **Tam Sayı** | `int` | Ondalık kısmı olmayan sayılar |
| **Ondalıklı Sayı** | `float` | Ondalık kısmı olan sayılar |

---

### Karakter Dizisi (String - str)

Karakter dizileri, tırnak içinde yazılan her türlü metni ifade eder. Harfler, rakamlar, boşluklar ve özel karakterler içerebilir.

```python
# String (Karakter Dizisi) veri tipi
# Tırnak içinde yazılan her değer karakter dizisidir
# Harfler, rakamlar, boşluklar ve özel karakterler içerebilir
isim = "Züber Doğan"  # Karakter dizisi değişkene atanıyor
print(isim)  # Değişkenin değerini yazdırır: Züber Doğan
print(type(isim))  # Veri tipini gösterir: <class 'str'>
```

### Tam Sayı (Integer - int)

Tam sayılar, ondalık kısmı olmayan pozitif veya negatif sayılardır. Python'da tam sayıların boyutu konusunda pratik bir sınır yoktur.

```python
# Integer (Tam Sayı) veri tipi
# Ondalık kısmı olmayan sayılar integer olarak kabul edilir
# Negatif veya pozitif olabilir, boyut sınırı pratik olarak yoktur
yil = 1978  # Tam sayı değişkene atanıyor
print(yil)  # Değerini yazdırır: 1978
print(type(yil))  # Veri tipini gösterir: <class 'int'>
```

### Ondalıklı Sayı (Float)

Ondalıklı sayılar, kesirli değerleri temsil eder. Python'da ondalık ayırıcı olarak nokta (`.`) kullanılır, virgül değil.

```python
# Float (Ondalıklı Sayı) veri tipi
# Nokta ile ayrılmış ondalık sayılar float olarak kabul edilir
# Python'da ondalık ayırıcı nokta (.) kullanılır, virgül değil
kilo = 73.3  # Ondalıklı sayı değişkene atanıyor
print(kilo)  # Değerini yazdırır: 73.3
print(type(kilo))  # Veri tipini gösterir: <class 'float'>
```

# **1.4 type() ve len() Fonksiyonları**

Python'da gömülü (built-in) fonksiyonlar, herhangi bir kütüphane içe aktarmadan doğrudan kullanabileceğimiz hazır fonksiyonlardır.

---

### type() Fonksiyonu

`type()` fonksiyonu, kendisine verilen değerin veya değişkenin **veri tipini** döndürür. Bu fonksiyon, özellikle hata ayıklama (debugging) sürecinde ve veri tiplerini kontrol etmek istediğimizde çok faydalıdır.

---

### len() Fonksiyonu

`len()` fonksiyonu, karakter dizilerinin **uzunluğunu** (karakter sayısını) döndürür. Ayrıca listeler, demetler ve sözlükler gibi veri yapılarının eleman sayısını öğrenmek için de kullanılır.

```python
# type() fonksiyonu - Veri tipini belirler ve döndürür
# Herhangi bir değerin hangi tipte olduğunu öğrenmek için kullanılır
# Hata ayıklama sürecinde çok faydalıdır
print(type("Merhaba"))  # str - Karakter dizisi (tırnak içinde olduğu için)
print(type(42))         # int - Tam sayı (ondalık kısmı yok)
print(type(3.14))       # float - Ondalıklı sayı (nokta ile ayrılmış)
```

```python
# len() fonksiyonu - Karakter dizisinin uzunluğunu (karakter sayısını) verir
# Boşluklar ve özel karakterler de sayılır
# Yalnızca karakter dizileri ve dizilere (list, tuple) uygulanabilir
metin = "firat_ozgul_1980"  # 16 karakterlik bir karakter dizisi
print(len(metin))  # 16 - Boşluk ve alt çizgi dahil tüm karakterler sayılır
```

```python
# Sayıların basamak sayısını bulmak için len() kullanımı
# Dikkat: len() doğrudan sayılara uygulanamaz
# Önce sayı str() ile karakter dizisine çevrilmeli, sonra len() uygulanmalı
sayi = 12343423432  # Bir tam sayı
print(len(str(sayi)))  # 11 - Sayı önce string'e çevrildi, sonra uzunluğu hesaplandı
```

# **1.5 print() Fonksiyonunun Özel Parametreleri**

`print()` fonksiyonu son derece güçlü bir araçtır ve davranışını özelleştirmemizi sağlayan **isimli parametreler** içerir. Bu parametreler, fonksiyonda görünmese bile her zaman oradadır ve varsayılan değerlere sahiptir.

---

### sep Parametresi (Separator - Ayırıcı)

`sep` parametresi, `print()` fonksiyonuna verilen parametrelerin arasına hangi karakterin yerleştirileceğini belirler. Varsayılan değeri bir adet boşluk karakteridir (`" "`).

---

### end Parametresi

`end` parametresi, `print()` fonksiyonunun çıktısının sonuna hangi karakterin ekleneceğini belirler. Varsayılan değeri yeni satır karakteridir (`"\n"`).

```python
# sep parametresi - Parametreler arası ayırıcı karakteri belirler
# Varsayılan değeri bir boşluk karakteridir (" ")
# sep="istenilen_karakter" şeklinde özelleştirilebilir
print("Adana", "Mersin", sep="-")  # Çıktı: Adana-Mersin (boşluk yerine tire)
```

```python
# sep parametresi ile yeni satır karakteri kullanımı
# \n özel karakteri yeni satır anlamına gelir
# Her parametre ayrı satırda yazdırılır
print("Adana", "Mersin", "İstanbul", sep="\n")  # Her şehir ayrı satırda
```

```python
# sep parametresi ile URL birleştirme örneği
# sep="" (boş karakter dizisi) parametreler arasına hiçbir şey koymaz
# Varsayılan boşluk karakteri atlanarak parametreler birleştirilir
print("http://", "www.", "python.", "org", sep="")  # Çıktı: http://www.python.org
```

```python
# end parametresi - Satır sonuna eklenecek karakteri belirler
# Varsayılan değeri yeni satır karakteridir (\n)
# end="istenilen_karakter" şeklinde özelleştirilebilir
print("Bugün günlerden Salı", end=".\n")  # Satır sonuna nokta ve yeni satır eklenir
```

```python
# end parametresi ile aynı satırda yazdırma
# end="" kullanıldığında bir sonraki print aynı satırda devam eder
print("Python ", end="")  # Satır sonu yok, bir sonraki çıktı aynı satırda
print("Programlama")  # Bu çıktı yukarıdakinin devamı olarak görünür
```

```python
# sep ve end parametrelerini birlikte kullanma
# Her iki parametre de aynı print() içinde kullanılabilir
# sep parametrelerin arasına, end ise en sona karakter ekler
print("T", "C", sep=".", end="\n\n")  # Çıktı: T.C ve iki satır boşluk
```

# **1.6 Kaçış Dizileri (Escape Sequences)**

Kaçış dizileri, karakter dizileri içinde özel anlam taşıyan karakter kombinasyonlarıdır. Ters eğik çizgi (`\`) ile başlarlar ve normal yollarla yazılamayan karakterleri temsil ederler.

| Kaçış Dizisi | Anlamı |
|--------------|--------|
| `\n` | Yeni satır (newline) |
| `\t` | Sekme (tab) |
| `\\` | Ters eğik çizgi |
| `\'` | Tek tırnak |
| `\"` | Çift tırnak |

```python
# \n kaçış dizisi - Yeni satır karakteri
# Karakter dizisini bulunduğu noktadan böler ve geri kalanı alt satıra geçirir
print("Birinci satır\nİkinci satır\nÜçüncü satır")  # Her \n'de alt satıra geçer
```

```python
# \t kaçış dizisi - Sekme (tab) karakteri
# Metinde belirli bir boşluk bırakır, hizalama için kullanışlıdır
print("İsim\tYaş\tŞehir")  # Sütunlar arasında tab boşluğu
print("Ali\t25\tAnkara")  # Veriler hizalı görünür
```

```python
# \' ve \" kaçış dizileri - Tırnak karakterlerini yazdırma
# Aynı tip tırnak içinde tırnak karakteri kullanmak gerektiğinde
print('İstanbul\'un tarihi')  # Tek tırnak içinde kesme işareti
print("\"Python\" bir programlama dilidir")  # Çift tırnak içinde çift tırnak
```
