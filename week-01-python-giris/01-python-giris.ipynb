# **1. Python'a Giriş**

Bu derste Python programlama diline giriş yapacağız. Print fonksiyonu, tırnak kullanımı ve temel veri tiplerini öğreneceğiz.

## **1.1 Print Fonksiyonu**

Python'da ekrana çıktı vermek için `print()` fonksiyonu kullanılır.

```python
# Ekrana metin yazdırma
print("Merhaba Dünya!")
```

```python
# Birden fazla değer yazdırma
print("Python", "Programlama", "Dersi")
```

## **1.2 Tırnak Kullanımı**

Python'da metin (string) tanımlamak için tek tırnak, çift tırnak veya üç tırnak kullanılabilir.

```python
# Tek tırnak kullanımı
print('Merhaba Python')
```

```python
# Çift tırnak içinde tek tırnak
print("yeni başlangıçlar'güzeldir'")
```

```python
# Tek tırnak içinde çift tırnak
print('Python programlama dilinin adı "piton" yılanından gelmez')
```

```python
# Üç tırnak - çok satırlı metin
print("""Bu bir
çok satırlı
metin örneğidir.""")
```

## **1.3 Veri Tipleri**

Python'da temel veri tipleri:
- **string (str)**: Karakter dizisi
- **integer (int)**: Tam sayı
- **float**: Ondalıklı sayı

```python
# String (Karakter Dizisi)
isim = "Züber Doğan"
print(isim)
print(type(isim))
```

```python
# Integer (Tam Sayı)
yil = 1978
print(yil)
print(type(yil))
```

```python
# Float (Ondalıklı Sayı)
kilo = 73.3
print(kilo)
print(type(kilo))
```

## **1.4 type() ve len() Fonksiyonları**

```python
# type() - Veri tipini gösterir
print(type("Merhaba"))  # str
print(type(42))         # int
print(type(3.14))       # float
```

```python
# len() - Karakter sayısını verir
metin = "firat_ozgul_1980"
print(len(metin))  # 16 karakter
```

```python
# Sayıları karakter dizisine çevirip uzunluğunu bulma
sayi = 12343423432
print(len(str(sayi)))  # 11 basamak
```

## **1.5 Print Özel Parametreleri**

```python
# sep parametresi - değerler arasına ayraç koyar
print("Adana", "Mersin", sep="-")
```

```python
# Alt satıra geçme
print("Adana", "Mersin", sep="\n")
```

```python
# end parametresi - satır sonuna karakter ekler
print("Bugün günlerden Salı", end=".\n")
```
