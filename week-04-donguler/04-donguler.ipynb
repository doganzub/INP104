# **4.2 while DÃ¶ngÃ¼sÃ¼**

`while` kelimesi Ä°ngilizce'de "... iken, ... olduÄŸu sÃ¼rece" anlamÄ±na gelir.

## SÃ¶z Dizimi

```python
while koÅŸul:
    # koÅŸul True olduÄŸu sÃ¼rece bu blok tekrarlanÄ±r
```

**Kurallar:**
- `while` satÄ±rÄ±nÄ±n sonunda `:` olmalÄ±
- DÃ¶ngÃ¼ bloÄŸu 4 boÅŸluk girintili yazÄ±lmalÄ±
- KoÅŸul `False` olduÄŸunda dÃ¶ngÃ¼ sona erer

````python
# while dÃ¶ngÃ¼sÃ¼ - Temel Ã¶rnek
# KoÅŸul doÄŸru olduÄŸu sÃ¼rece blok tekrarlanÄ±r

a = 1  # SayaÃ§ deÄŸiÅŸkeni baÅŸlangÄ±Ã§ deÄŸeri

# a deÄŸiÅŸkeni 10'dan kÃ¼Ã§Ã¼k olduÄŸu sÃ¼rece dÃ¶ngÃ¼ye devam et
while a < 10:
    print(a)  # Mevcut deÄŸeri ekrana yazdÄ±r
    a += 1    # SayacÄ± 1 artÄ±r (a = a + 1 ile aynÄ±)
              # Bu satÄ±r olmazsa SONSUZ DÃ–NGÃœ oluÅŸur!

print("DÃ¶ngÃ¼ bitti")  # KoÅŸul False olunca buraya gelir
````

## Sonsuz DÃ¶ngÃ¼ Tehlikesi

EÄŸer dÃ¶ngÃ¼ koÅŸulunu deÄŸiÅŸtirecek bir kod yazmazsak, program **sonsuz dÃ¶ngÃ¼ye** girer:

```python
# YANLIÅ - Sonsuz dÃ¶ngÃ¼!
a = 1
while a < 10:
    print("Bu satÄ±r sonsuza kadar yazÄ±lÄ±r!")
    # a hiÃ§ artmÄ±yor, koÅŸul hep True kalÄ±yor
```

**Sonsuz dÃ¶ngÃ¼den Ã§Ä±kmak iÃ§in:** `Ctrl+C` veya `Ctrl+Z`

````python
# while dÃ¶ngÃ¼sÃ¼ - KoÅŸul deÄŸiÅŸikliÄŸinin Ã¶nemi
# += operatÃ¶rÃ¼ ile sayaÃ§ artÄ±rma

tekrar = 1  # DÃ¶ngÃ¼ sayacÄ±

# tekrar deÄŸiÅŸkeni 3'e eÅŸit veya 3'ten kÃ¼Ã§Ã¼k olduÄŸu sÃ¼rece dÃ¶ngÃ¼ye devam et
while tekrar <= 3:
    print(f"DÃ¶ngÃ¼ #{tekrar}")  # KaÃ§Ä±ncÄ± dÃ¶ngÃ¼de olduÄŸumuzu gÃ¶ster
    tekrar += 1                 # SayacÄ± 1 artÄ±r
    # tekrar 4 olduÄŸunda tekrar <= 3 koÅŸulu False olur ve dÃ¶ngÃ¼ biter

print(f"DÃ¶ngÃ¼ sona erdi. Son deÄŸer: {tekrar}")  # 4
````

````python
# while dÃ¶ngÃ¼sÃ¼ - Arka planda neler oluyor?
# bool() fonksiyonu ile koÅŸulun doÄŸruluk deÄŸerini gÃ¶relim

sayac = 1  # BaÅŸlangÄ±Ã§ deÄŸeri

while sayac <= 3:
    print(f"sayac deÄŸeri: {sayac}")  # Mevcut sayaÃ§ deÄŸeri
    
    sayac += 1  # SayacÄ± artÄ±r
    
    # KoÅŸulun bool deÄŸerini kontrol et
    # True ise dÃ¶ngÃ¼ devam eder, False ise dÃ¶ngÃ¼ biter
    print(f"sayac <= 3 koÅŸulunun bool deÄŸeri: {bool(sayac <= 3)}")
    print("---")
````

# **4.3 while True YapÄ±sÄ± ve break**

`while True` ifadesi, "aksi belirtilmediÄŸi sÃ¼rece Ã§alÄ±ÅŸmaya devam et" anlamÄ±na gelir.

```python
while True:
    # Bu blok sonsuza kadar Ã§alÄ±ÅŸÄ±r
    # Durdurmak iÃ§in break gerekli!
```

Bu yapÄ± genellikle `break` deyimi ile birlikte kullanÄ±lÄ±r.

````python
# while True ve break - KullanÄ±cÄ± Ã§Ä±kana kadar devam et
# break deyimi dÃ¶ngÃ¼yÃ¼ "kÄ±rar" ve sonlandÄ±rÄ±r

while True:  # Aksi belirtilmediÄŸi sÃ¼rece Ã§alÄ±ÅŸ
    soru = input("NasÄ±lsÄ±nÄ±z? (Ã‡Ä±kmak iÃ§in 'q' yazÄ±n): ")
    
    if soru == "q":  # KullanÄ±cÄ± 'q' tuÅŸuna bastÄ±ysa
        print("Ã‡Ä±kÄ±lÄ±yor...")
        break  # DÃ¶ngÃ¼yÃ¼ kÄ±r ve Ã§Ä±k
    
    print(f"CevabÄ±nÄ±z: {soru}")

print("Program sona erdi.")
````

````python
# while True ile hesap makinesi Ã¶rneÄŸi
# KullanÄ±cÄ± 'q' girene kadar Ã§alÄ±ÅŸÄ±r

print("""
Basit Hesap Makinesi
--------------------
(1) Toplama
(2) Ã‡Ä±karma
(q) Ã‡Ä±kÄ±ÅŸ
""")

while True:  # Sonsuz dÃ¶ngÃ¼ baÅŸlat
    secim = input("Ä°ÅŸlem seÃ§in: ")  # KullanÄ±cÄ±dan seÃ§im al
    
    if secim == "q":  # Ã‡Ä±kÄ±ÅŸ kontrolÃ¼
        print("Hesap makinesi kapatÄ±lÄ±yor...")
        break  # DÃ¶ngÃ¼den Ã§Ä±k
    
    elif secim == "1":  # Toplama iÅŸlemi
        # input() string dÃ¶ndÃ¼rÃ¼r, int() ile sayÄ±ya Ã§evir
        sayi1 = int(input("Ä°lk sayÄ±: "))
        sayi2 = int(input("Ä°kinci sayÄ±: "))
        print(f"{sayi1} + {sayi2} = {sayi1 + sayi2}")
    
    elif secim == "2":  # Ã‡Ä±karma iÅŸlemi
        sayi1 = int(input("Ä°lk sayÄ±: "))
        sayi2 = int(input("Ä°kinci sayÄ±: "))
        print(f"{sayi1} - {sayi2} = {sayi1 - sayi2}")
    
    else:  # GeÃ§ersiz seÃ§im
        print("GeÃ§ersiz seÃ§im!")
````

# **4.4 for DÃ¶ngÃ¼sÃ¼**

`for` dÃ¶ngÃ¼sÃ¼, bir koleksiyon (karakter dizisi, liste vb.) Ã¼zerinde **eleman eleman** dolaÅŸÄ±r.

## SÃ¶z Dizimi

```python
for deÄŸiÅŸken_adÄ± in koleksiyon:
    # her eleman iÃ§in bu blok Ã§alÄ±ÅŸÄ±r
```

**TÃ¼rkÃ§e karÅŸÄ±lÄ±ÄŸÄ±:**

```
koleksiyon iÃ§indeki her bir Ã¶ÄŸeyi deÄŸiÅŸken_adÄ± olarak adlandÄ±r:
    ve bu Ã¶ÄŸelerle bir iÅŸlem yap.
```

````python
# for dÃ¶ngÃ¼sÃ¼ - Karakter dizisi Ã¼zerinde dÃ¶ngÃ¼
# Her harf tek tek ele alÄ±nÄ±r

isim = "Python"  # Ãœzerinde dÃ¶ngÃ¼ kurulacak karakter dizisi

# isim deÄŸiÅŸkeni iÃ§indeki her bir karakteri 'harf' olarak adlandÄ±r
for harf in isim:
    print(harf)  # Her turda bir harf yazdÄ±rÄ±lÄ±r

# Ã‡Ä±ktÄ±: P, y, t, h, o, n (her biri ayrÄ± satÄ±rda)
````

````python
# for dÃ¶ngÃ¼sÃ¼ - TÃ¼rkÃ§e karakterler
# Python TÃ¼rkÃ§e karakterleri tam destekler

tr_harfler = "ÅŸÃ§Ã¶ÄŸÃ¼Ä°Ä±"  # TÃ¼rkÃ§eye Ã¶zgÃ¼ harfler

# Her TÃ¼rkÃ§e harfi tek tek yazdÄ±r
for harf in tr_harfler:
    print(harf)
````

````python
# for dÃ¶ngÃ¼sÃ¼ - SayÄ±larla iÅŸlem yapma
# String sayÄ±larÄ± int()'e Ã§evirerek matematiksel iÅŸlem

sayilar = "123456789"  # String olarak tanÄ±mlÄ± sayÄ±lar

# Her bir karakteri sayÄ±ya Ã§evirip 2 ile Ã§arp
for sayi in sayilar:
    # sayi deÄŸiÅŸkeni string, int() ile sayÄ±ya Ã§evir
    sonuc = int(sayi) * 2
    print(f"{sayi} x 2 = {sonuc}")
````

# **4.5 range() Fonksiyonu**

`range` kelimesi Ä°ngilizce'de "aralÄ±k" anlamÄ±na gelir. Bu fonksiyon, belirli bir aralÄ±kta sayÄ±lar Ã¼retir.

## KullanÄ±m Åekilleri

| SÃ¶z Dizimi | AÃ§Ä±klama | Ã–rnek |
|------------|----------|-------|
| `range(stop)` | 0'dan stop-1'e kadar | `range(5)` â†’ 0,1,2,3,4 |
| `range(start, stop)` | start'tan stop-1'e kadar | `range(1,6)` â†’ 1,2,3,4,5 |
| `range(start, stop, step)` | start'tan stop-1'e, step kadar atlayarak | `range(0,10,2)` â†’ 0,2,4,6,8 |

**Ã–nemli:** `stop` deÄŸeri sonuca **dahil deÄŸildir**!

````python
# range(stop) - 0'dan stop-1'e kadar
# Tek parametre verildiÄŸinde baÅŸlangÄ±Ã§ 0 kabul edilir

print("range(5) Ã§Ä±ktÄ±sÄ±:")
for i in range(5):  # 0, 1, 2, 3, 4 (5 dahil deÄŸil!)
    print(i, end=" ")  # end=" " ile aynÄ± satÄ±rda yazdÄ±r

print()  # Yeni satÄ±r
````

````python
# range(start, stop) - BaÅŸlangÄ±Ã§ ve bitiÅŸ deÄŸeri
# start dahil, stop dahil deÄŸil

print("range(1, 6) Ã§Ä±ktÄ±sÄ±:")
for i in range(1, 6):  # 1, 2, 3, 4, 5 (6 dahil deÄŸil!)
    print(i, end=" ")

print()

print("range(3, 10) Ã§Ä±ktÄ±sÄ±:")
for i in range(3, 10):  # 3'ten 9'a kadar
    print(i, end=" ")
````

````python
# range(start, stop, step) - AdÄ±m deÄŸeri ile
# step: kaÃ§ar kaÃ§ar atlayacaÄŸÄ±nÄ± belirler

print("range(0, 10, 2) - Ã‡ift sayÄ±lar:")
for i in range(0, 10, 2):  # 0'dan 10'a, 2'ÅŸer atlayarak
    print(i, end=" ")  # 0, 2, 4, 6, 8

print()

print("range(1, 10, 2) - Tek sayÄ±lar:")
for i in range(1, 10, 2):  # 1'den 10'a, 2'ÅŸer atlayarak
    print(i, end=" ")  # 1, 3, 5, 7, 9

print()

print("range(0, 100, 10) - 10'ar atlayarak:")
for i in range(0, 100, 10):  # 0'dan 100'e, 10'ar atlayarak
    print(i, end=" ")  # 0, 10, 20, ..., 90
````

````python
# range() ile tersten sayma
# Negatif step deÄŸeri kullanÄ±lÄ±r

print("range(10, 0, -1) - Geriye doÄŸru sayma:")
for i in range(10, 0, -1):  # 10'dan 1'e, -1 adÄ±mla
    print(i, end=" ")  # 10, 9, 8, ..., 1

print()

print("range(10, 0, -3) - 3'er geri:")
for i in range(10, 0, -3):  # 10'dan 1'e, 3'er geri
    print(i, end=" ")  # 10, 7, 4, 1
````

# **4.6 DÃ¶ngÃ¼ Kontrol Deyimleri**

DÃ¶ngÃ¼lerin akÄ±ÅŸÄ±nÄ± kontrol etmek iÃ§in Ã¼Ã§ Ã¶nemli deyim kullanÄ±lÄ±r:

| Deyim | AnlamÄ± | Ä°ÅŸlevi |
|-------|--------|--------|
| `break` | KÄ±r, durdur | DÃ¶ngÃ¼yÃ¼ tamamen sonlandÄ±rÄ±r |
| `continue` | Devam et | Mevcut adÄ±mÄ± atlar, sonrakine geÃ§er |
| `pass` | GeÃ§ | HiÃ§bir ÅŸey yapma, yer tutucu |

````python
# break deyimi - DÃ¶ngÃ¼yÃ¼ tamamen sonlandÄ±r
# Belirli bir koÅŸul saÄŸlandÄ±ÄŸÄ±nda dÃ¶ngÃ¼den Ã§Ä±k

print("break Ã¶rneÄŸi:")
for i in range(10):  # 0'dan 9'a kadar
    if i == 5:               # i deÄŸeri 5'e eÅŸit olduÄŸunda
        print("5 bulundu, dÃ¶ngÃ¼ sonlandÄ±rÄ±lÄ±yor!")
        break                # DÃ¶ngÃ¼den tamamen Ã§Ä±k
    print(i, end=" ")        # Sadece 0, 1, 2, 3, 4 yazdÄ±rÄ±lÄ±r

print("\nDÃ¶ngÃ¼ bitti.")
````

````python
# continue deyimi - Mevcut adÄ±mÄ± atla, sonrakine geÃ§
# Belirli deÄŸerler iÃ§in iÅŸlem yapmadan devam et

print("continue Ã¶rneÄŸi - Ã‡ift sayÄ±larÄ± atla:")
for i in range(10):  # 0'dan 9'a kadar
    if i % 2 == 0:           # i deÄŸeri Ã§ift sayÄ± ise
        continue             # Bu turu atla, sonrakine geÃ§
    print(i, end=" ")        # Sadece tek sayÄ±lar: 1, 3, 5, 7, 9
````

````python
# continue Ã¶rneÄŸi - ÃœÃ§ haneli olmayan sayÄ±larÄ± atla
# KullanÄ±cÄ± giriÅŸi kontrolÃ¼

while True:
    s = input("Bir sayÄ± girin (Ã§Ä±kmak iÃ§in 'q'): ")
    
    if s == "q":       # Ã‡Ä±kÄ±ÅŸ kontrolÃ¼
        break
    
    if len(s) <= 3:    # 3 haneli veya daha az ise
        continue       # UyarÄ±yÄ± atla, baÅŸa dÃ¶n
    
    # Bu satÄ±r sadece 3'ten fazla haneli sayÄ±lar iÃ§in Ã§alÄ±ÅŸÄ±r
    print("En fazla Ã¼Ã§ haneli bir sayÄ± girebilirsiniz.")
````

````python
# pass deyimi - HiÃ§bir ÅŸey yapma (yer tutucu)
# KoÅŸul saÄŸlandÄ±ÄŸÄ±nda hiÃ§bir iÅŸlem yapmadan geÃ§

print("pass Ã¶rneÄŸi - Negatif sayÄ±larÄ± gÃ¶rmezden gel:")
for i in range(-3, 4):  # -3'ten 3'e kadar
    if i < 0:                # Negatif sayÄ±lar iÃ§in
        pass                 # HiÃ§bir ÅŸey yapma
    else:
        print(i, end=" ")    # Sadece 0, 1, 2, 3 yazdÄ±rÄ±lÄ±r
````

````python
# pass deyimi - Taslak kod iÃ§in yer tutucu
# HenÃ¼z yazÄ±lmamÄ±ÅŸ kod bloklarÄ± iÃ§in kullanÄ±lÄ±r

sayi = 10

if sayi > 0:
    print("SayÄ± pozitif")
elif sayi < 0:
    print("SayÄ± negatif")
else:
    pass  # SayÄ± 0 olduÄŸunda ne yapÄ±lacaÄŸÄ±na henÃ¼z karar verilmedi
          # pass olmadan bu blok boÅŸ kalamaz, SyntaxError verir
````

# **4.7 DÃ¶ngÃ¼ ile else KullanÄ±mÄ±**

Python'da `else` deyimi dÃ¶ngÃ¼lerle birlikte de kullanÄ±labilir. Bu, diÄŸer programlama dillerinde nadir gÃ¶rÃ¼len bir Ã¶zelliktir.

**Kurallar:**
- DÃ¶ngÃ¼ `break` ile **sonlandÄ±rÄ±lmadÄ±ysa** â†’ `else` bloÄŸu **Ã§alÄ±ÅŸÄ±r**
- DÃ¶ngÃ¼ `break` ile **sonlandÄ±rÄ±ldÄ±ysa** â†’ `else` bloÄŸu **Ã§alÄ±ÅŸmaz**

````python
# for-else yapÄ±sÄ± - break Ã§alÄ±ÅŸmadÄ±ÄŸÄ±nda else Ã§alÄ±ÅŸÄ±r
# Aramada bulunmama durumunu kontrol etmek iÃ§in kullanÄ±ÅŸlÄ±

print("=== break Ã§alÄ±ÅŸmayan Ã¶rnek ===")
for i in range(5):  # 0'dan 4'e kadar
    print(i, end=" ")
else:  # DÃ¶ngÃ¼ normal ÅŸekilde (break olmadan) bittiyse
    print("\nDÃ¶ngÃ¼ break olmadan tamamlandÄ±, else Ã§alÄ±ÅŸtÄ±.")
````

````python
# for-else yapÄ±sÄ± - break Ã§alÄ±ÅŸtÄ±ÄŸÄ±nda else Ã§alÄ±ÅŸmaz

print("=== break Ã§alÄ±ÅŸan Ã¶rnek ===")
for i in range(10):
    print(i, end=" ")
    if i == 3:       # 3'e ulaÅŸÄ±nca
        print("\nbreak ile Ã§Ä±kÄ±ldÄ±")
        break        # DÃ¶ngÃ¼yÃ¼ kÄ±r
else:  # break Ã§alÄ±ÅŸtÄ±ÄŸÄ± iÃ§in burasÄ± Ã§alÄ±ÅŸmaz!
    print("Bu satÄ±r gÃ¶rÃ¼nmez Ã§Ã¼nkÃ¼ break Ã§alÄ±ÅŸtÄ±.")
````

````python
# Pratik Ã¶rnek - Karakter arama
# Bir metinde belirli bir karakterin olup olmadÄ±ÄŸÄ±nÄ± kontrol et

metin = "Merhaba DÃ¼nya"
aranan = "x"  # Aranacak karakter

for harf in metin:
    if harf == aranan:  # Karakter bulunduysa
        print(f"'{aranan}' harfi bulundu!")
        break           # DÃ¶ngÃ¼den Ã§Ä±k
else:  # break Ã§alÄ±ÅŸmadÄ±ysa = karakter bulunamadÄ±ysa
    print(f"'{aranan}' harfi bu metinde yok.")
````

# **4.8 Ä°Ã§ Ä°Ã§e DÃ¶ngÃ¼ler (Nested Loops)**

Bir dÃ¶ngÃ¼ iÃ§inde baÅŸka bir dÃ¶ngÃ¼ kullanÄ±labilir. DÄ±ÅŸ dÃ¶ngÃ¼nÃ¼n **her bir turu** iÃ§in iÃ§ dÃ¶ngÃ¼ **tamamen** Ã§alÄ±ÅŸÄ±r.

````python
# Ä°Ã§ iÃ§e dÃ¶ngÃ¼ - Temel Ã¶rnek
# DÄ±ÅŸ dÃ¶ngÃ¼nÃ¼n her turu iÃ§in iÃ§ dÃ¶ngÃ¼ tamamen Ã§alÄ±ÅŸÄ±r

for dis in range(3):  # DÄ±ÅŸ dÃ¶ngÃ¼: 0, 1, 2
    print(f"DÄ±ÅŸ dÃ¶ngÃ¼: {dis}")
    
    for ic in range(2):  # Ä°Ã§ dÃ¶ngÃ¼: 0, 1
        print(f"  Ä°Ã§ dÃ¶ngÃ¼: {ic}")  # Girinti ile iÃ§ dÃ¶ngÃ¼ olduÄŸunu gÃ¶ster
    
    print("---")  # Her dÄ±ÅŸ dÃ¶ngÃ¼ turu sonunda ayraÃ§
````

````python
# Ä°Ã§ iÃ§e dÃ¶ngÃ¼ - Ã‡arpÄ±m tablosu
# 5x5 Ã§arpÄ±m tablosu oluÅŸturma

print("Ã‡arpÄ±m Tablosu (5x5)")
print("=" * 30)

for i in range(1, 6):         # SatÄ±rlar: 1, 2, 3, 4, 5
    for j in range(1, 6):     # SÃ¼tunlar: 1, 2, 3, 4, 5
        # end="\t" ile tab boÅŸluÄŸu bÄ±rak, satÄ±r sonuna geÃ§me
        print(f"{i}x{j}={i*j}", end="\t")
    print()  # Her satÄ±r sonunda yeni satÄ±ra geÃ§
````

````python
# Ä°Ã§ iÃ§e while dÃ¶ngÃ¼sÃ¼
# KontrollÃ¼ iÃ§ dÃ¶ngÃ¼ Ã¶rneÄŸi

dis = 0  # DÄ±ÅŸ dÃ¶ngÃ¼ sayacÄ±

while dis < 3:  # DÄ±ÅŸ dÃ¶ngÃ¼
    print(f"DÄ±ÅŸ: {dis}")
    
    ic = 0  # Ä°Ã§ dÃ¶ngÃ¼ sayacÄ± (her dÄ±ÅŸ turda sÄ±fÄ±rla!)
    while ic < 2:  # Ä°Ã§ dÃ¶ngÃ¼
        print(f"  Ä°Ã§: {ic}")
        ic += 1  # Ä°Ã§ sayacÄ± artÄ±r
    
    dis += 1  # DÄ±ÅŸ sayacÄ± artÄ±r

print("TamamlandÄ±.")
````

# **4.9 Pratik Ã–rnekler**

````python
# Pratik Ã–rnek 1: Ã‡ift SayÄ±larÄ± Bulma (while ile)
# 1-100 arasÄ±ndaki Ã§ift sayÄ±larÄ± yazdÄ±r

print("1-100 arasÄ± Ã§ift sayÄ±lar:")
a = 0  # SayaÃ§ deÄŸiÅŸkeni

while a < 100:  # 100'e kadar devam et
    a += 1      # Ã–nce artÄ±r
    
    if a % 2 == 0:    # Ã‡ift sayÄ± mÄ±? (2'ye tam bÃ¶lÃ¼nÃ¼yor mu?)
        print(a, end=" ")
````

````python
# Pratik Ã–rnek 2: Parola DoÄŸrulama (3 deneme hakkÄ±)
# for + else yapÄ±sÄ± ile gÃ¼venli giriÅŸ sistemi

dogru_parola = "python123"  # Sistem parolasÄ±

for deneme in range(3):  # 3 deneme hakkÄ± (0, 1, 2)
    parola = input(f"Parola girin (Deneme {deneme + 1}/3): ")
    
    if not parola:  # BoÅŸ girdi kontrolÃ¼
        print("Parola boÅŸ bÄ±rakÄ±lamaz!")
    
    elif parola == dogru_parola:  # DoÄŸru parola
        print("GiriÅŸ baÅŸarÄ±lÄ±!")
        break  # DÃ¶ngÃ¼den Ã§Ä±k
    
    elif len(parola) not in range(3, 9):  # Uzunluk kontrolÃ¼
        print("Parola 3-8 karakter arasÄ±nda olmalÄ±!")
    
    else:  # YanlÄ±ÅŸ parola
        print("YanlÄ±ÅŸ parola!")

else:  # break Ã§alÄ±ÅŸmadÄ±ysa = 3 deneme bitti, giriÅŸ baÅŸarÄ±sÄ±z
    print("3 deneme hakkÄ±nÄ±z doldu! Hesap kilitlendi.")
````

````python
# Pratik Ã–rnek 3: TÃ¼rkÃ§e Karakter KontrolÃ¼
# Parolada TÃ¼rkÃ§e karakter kullanÄ±lmasÄ±nÄ± engelle

tr_harfler = "ÅŸÃ§Ã¶ÄŸÃ¼Ä°Ä±ÅÃ‡Ã–ÄÃœ"  # TÃ¼rkÃ§e karakterler

parola = input("Yeni parola belirleyin: ")

# Paroladaki her karakteri kontrol et
for karakter in parola:
    if karakter in tr_harfler:  # TÃ¼rkÃ§e karakter varsa
        print(f"Hata: '{karakter}' - TÃ¼rkÃ§e karakter kullanÄ±lamaz!")
        break
else:  # HiÃ§ TÃ¼rkÃ§e karakter yoksa
    print(f"Parola kabul edildi: {parola}")
````
