# **4. Döngüler**

Bu derste Python'da döngü yapılarını (while, for) ve döngü kontrol ifadelerini öğreneceğiz.

## **4.1 while Döngüsü**

while döngüsü, belirli bir koşul doğru olduğu sürece kod bloğunu tekrarlar.

```python
# Basit while döngüsü
sayac = 0

while sayac < 5:
    print(f"Sayaç: {sayac}")
    sayac = sayac + 1

print("Döngü bitti")
```

```python
# += operatörü ile kısa yazım
sayac = 0

while sayac < 10:
    print(sayac)
    sayac += 1  # sayac = sayac + 1 ile aynı
```

## **4.2 for Döngüsü**

for döngüsü, bir koleksiyon (string, liste, vb.) üzerinde iterasyon yapar.

```python
# String üzerinde döngü
isim = "Python"

for harf in isim:
    print(harf)
```

```python
# Türkçe karakterler üzerinde döngü
tr_harfler = "şçöğüİı"

for harf in tr_harfler:
    print(harf)
```

## **4.3 range() Fonksiyonu**

```python
# range(stop) - 0'dan stop-1'e kadar
for i in range(5):
    print(i)  # 0, 1, 2, 3, 4
```

```python
# range(start, stop) - start'tan stop-1'e kadar
for i in range(1, 6):
    print(i)  # 1, 2, 3, 4, 5
```

```python
# range(start, stop, step) - adım değeri ile
for i in range(0, 10, 2):
    print(i)  # 0, 2, 4, 6, 8 (çift sayılar)
```

## **4.4 break, continue, pass**

```python
# break - Döngüyü tamamen sonlandırır
for i in range(10):
    if i == 5:
        print("5 bulundu, döngü durduruluyor")
        break
    print(i)
```

```python
# continue - Mevcut adımı atla, sonraki adıma geç
for i in range(10):
    if i % 2 == 0:  # Çift sayıları atla
        continue
    print(i)  # Sadece tek sayılar
```

```python
# pass - Hiçbir şey yapma (yer tutucu)
for i in range(-5, 6):
    if i < 0:
        pass  # Negatif sayılarda hiçbir şey yapma
    else:
        print(i)
```

## **4.5 İç İçe Döngüler**

```python
# İç içe while döngüsü
d = 0
while d < 3:
    print(f"{d} - dış döngü")
    i = 0
    while i < 2:
        print(f"  {i} - iç döngü")
        i += 1
    d += 1
print("Çıkış")
```

```python
# Çarpım tablosu
for i in range(1, 6):
    for j in range(1, 6):
        print(f"{i}x{j}={i*j}", end="\t")
    print()  # Yeni satır
```

## **4.6 Pratik Örnek: Parola Doğrulama**

```python
# 3 deneme hakkı ile parola kontrolü
dogru_parola = "123"

for deneme in range(3):
    parola = input(f"Parola girin (Deneme {deneme + 1}/3): ")
    
    if not parola:
        print("Parola boş bırakılamaz!")
    elif parola == dogru_parola:
        print("Giriş başarılı!")
        break
    elif len(parola) not in range(3, 9):
        print("Parola 3-8 karakter olmalı!")
    else:
        print("Yanlış parola, tekrar deneyin")
else:
    print("3 deneme hakkınız doldu!")
```
