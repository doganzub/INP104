# **3.1 Koşullu Durumlar ve if Deyimi**

Programlama dillerinde **koşullu durumlar**, programın belirli koşullara göre farklı davranışlar sergilemesini sağlayan en önemli kavramlardan biridir.

Günlük hayattan örnekler:
- **E-posta girişi**: Kullanıcı adı ve parola doğruysa → hesaba erişim sağlanır, değilse → erişim engellenir
- **Sistem güncellemesi**: Kullanıcı "evet" derse → güncelleme başlar, "hayır" derse → işlem iptal edilir
- **Yaş kontrolü**: Kullanıcı 18 yaşından büyükse → programa alınır, değilse → uyarı gösterilir

Python'da koşullu durumları ifade etmek için üç deyim kullanılır:
1. `if` → "eğer"
2. `elif` → "eğer değilse ve eğer" (else if)
3. `else` → "aksi halde"

---

### if Deyiminin Söz Dizimi

```python
if koşul:
    # koşul doğruysa bu blok çalışır
```

**Önemli Kurallar:**
- `if` satırının sonunda `:` (iki nokta) olmalıdır
- Koşul bloğu **4 boşluk** girintili yazılmalıdır
- Girintileme Python'da zorunludur, boşluklar veya tab tutarlı olmalıdır

````python
# if deyimi - Basit koşul kontrolü
# Koşul doğruysa (True) blok içindeki kod çalışır
# Koşul yanlışsa (False) blok atlanır
n = 255  # Kontrol edilecek değişken

# n'nin 10'dan büyük olup olmadığını kontrol et
# > işareti "büyüktür" anlamına gelir
if n > 10:
    print("Sayı 10'dan büyüktür!")  # Bu satır girintili yazılmalı (4 boşluk)
````

````python
# Kullanıcıdan alınan değerle if kontrolü
# input() her zaman string döndürür, int() ile sayıya çeviriyoruz
sayi = int(input("Bir sayı giriniz: "))

# Her koşul ayrı bir if bloğunda - Python hepsini ayrı değerlendirir
if sayi > 10:
    print("Girdiğiniz sayı 10'dan büyüktür!")

if sayi < 10:
    print("Girdiğiniz sayı 10'dan küçüktür!")

if sayi == 10:  # == eşitlik kontrolü (= atama operatöründen farklı!)
    print("Girdiğiniz sayı 10'dur!")
````

# **3.2 Karşılaştırma Operatörleri**

Koşullu ifadelerde iki değeri karşılaştırmak için **karşılaştırma operatörleri** kullanılır. Bu operatörler her zaman `True` veya `False` (Boolean) değer döndürür.

| Operatör | Anlamı | Örnek | Sonuç |
|----------|--------|-------|-------|
| `>` | Büyüktür | `10 > 5` | `True` |
| `<` | Küçüktür | `10 < 5` | `False` |
| `>=` | Büyük eşittir | `10 >= 10` | `True` |
| `<=` | Küçük eşittir | `5 <= 10` | `True` |
| `==` | Eşittir | `5 == 5` | `True` |
| `!=` | Eşit değildir | `5 != 10` | `True` |

**Dikkat:** `=` (tek eşittir) **atama** operatörüdür, `==` (çift eşittir) **karşılaştırma** operatörüdür!

````python
# Karşılaştırma operatörleri - Her biri True veya False döndürür
a = 10  # İlk değer
b = 5   # İkinci değer

# == operatörü: İki değer eşit mi?
print(f"{a} == {b}:", a == b)   # False - 10, 5'e eşit değil

# != operatörü: İki değer farklı mı?
print(f"{a} != {b}:", a != b)   # True - 10, 5'ten farklı

# > operatörü: Sol taraf sağdan büyük mü?
print(f"{a} > {b}:", a > b)     # True - 10, 5'ten büyük

# < operatörü: Sol taraf sağdan küçük mü?
print(f"{a} < {b}:", a < b)     # False - 10, 5'ten küçük değil

# >= operatörü: Sol taraf sağdan büyük veya eşit mi?
print(f"{a} >= {b}:", a >= b)   # True - 10, 5'ten büyük

# <= operatörü: Sol taraf sağdan küçük veya eşit mi?
print(f"{a} <= {b}:", a <= b)   # False - 10, 5'e küçük veya eşit değil
````

````python
# = ve == arasındaki fark - ÇOK ÖNEMLİ!
# = → Değer atama (assignment)
# == → Eşitlik karşılaştırması (comparison)

x = 26       # x değişkenine 26 değerini ATA
print("x =", x)  # x = 26

# x'in 26'ya eşit olup olmadığını KONTROL ET
sonuc = x == 26  # Sonuç True veya False olacak
print("x == 26:", sonuc)  # True
````

# **3.3 if-else Yapısı**

`else` deyimi, `if` koşulu sağlanmadığında çalışacak alternatif bir blok tanımlar.

```python
if koşul:
    # koşul True ise bu blok çalışır
else:
    # koşul False ise bu blok çalışır
```

`else` deyiminin anlamı: **"Eğer yukarıdaki koşul(lar) gerçekleşmezse..."**

````python
# if-else yapısı - İki alternatif durum
# Koşul doğruysa if bloğu, yanlışsa else bloğu çalışır
yas = 15  # Kontrol edilecek yaş değeri

if yas >= 18:  # Yaş 18 veya üstü mü?
    print("Reşitsiniz")  # Bu blok yas >= 18 olduğunda çalışır
else:
    print("Reşit değilsiniz")  # Bu blok yas < 18 olduğunda çalışır
````

````python
# Parola kontrolü örneği
# if-else yapısıyla basit bir giriş sistemi
parola = input("Parola: ")  # Kullanıcıdan parola al

# Girilen parola doğru mu kontrol et
if parola == "12345678":  # == ile karşılaştırma
    print("Sisteme Hoşgeldiniz!")  # Doğru parola
else:
    print("Yanlış parola girdiniz!")  # Yanlış parola
````

# **3.4 if-elif-else Yapısı**

`elif` (else if), birden fazla koşulun sırayla kontrol edilmesi gerektiğinde kullanılır.

```python
if koşul_1:
    # koşul_1 True ise çalışır
elif koşul_2:
    # koşul_1 False VE koşul_2 True ise çalışır
elif koşul_3:
    # koşul_1 ve koşul_2 False VE koşul_3 True ise çalışır
else:
    # hiçbir koşul True değilse çalışır
```

---

### if vs elif Arasındaki Kritik Fark

| Art arda `if` blokları | `if-elif` yapısı |
|------------------------|------------------|
| Her `if` **ayrı ayrı** değerlendirilir | İlk doğru koşul bulunduğunda **durur** |
| Tüm doğru koşullar çalışır | Yalnızca **ilk doğru** koşul çalışır |
| Birbirinden bağımsız kontroller | Birbirine bağlı kontroller |

````python
# if-elif-else yapısı - Sınav puanı değerlendirme
# Sadece ilk doğru koşul çalışır, diğerleri atlanır
puan = int(input("Sınav puanınızı girin: "))

# Koşullar sırayla kontrol edilir
if puan >= 0 and puan <= 49:       # 0-49 arası
    print("Kaldınız")              # İlk koşul True ise burada durur
elif puan >= 50 and puan <= 64:    # 50-64 arası
    print("Geçer")                 # İkinci koşul True ise burada durur
elif puan >= 65 and puan <= 79:    # 65-79 arası
    print("Orta")
elif puan >= 80 and puan <= 89:    # 80-89 arası
    print("İyi")
elif puan >= 90 and puan <= 100:   # 90-100 arası
    print("Pekiyi")
else:                               # Hiçbiri değilse (geçersiz puan)
    print("Lütfen 0 ile 100 arasında puan giriniz")
````

````python
# if vs elif farkını göstermek için örnek
# Aynı değer (40) için farklı davranışlar

a = 40
print("=== Art arda if blokları ===")

# Her if ayrı değerlendirilir - İKİ sonuç verir!
if a < 100:
    print("Sayı 100'den küçük")  # Bu çalışır (40 < 100)

if a < 50:
    print("Sayı 50'den küçük")   # Bu DA çalışır (40 < 50)

print("\n=== if-elif yapısı ===")

# elif ile sadece İLK doğru koşul çalışır
if a < 100:
    print("Sayı 100'den küçük")  # Bu çalışır ve durur!
elif a < 50:
    print("Sayı 50'den küçük")   # Buraya HİÇ ulaşmaz
````

# **3.5 Mantıksal Operatörler**

Birden fazla koşulu birleştirmek için **mantıksal operatörler** kullanılır.

| Operatör | Anlamı | Açıklama |
|----------|--------|----------|
| `and` | Ve | Her iki koşul da `True` olmalı |
| `or` | Veya | En az bir koşul `True` olmalı |
| `not` | Değil | Koşulun tersini alır |

````python
# and operatörü - Her iki koşul da True olmalı
# Tüm koşullar sağlanmazsa blok çalışmaz
yas = 25     # Kullanıcının yaşı
gelir = 5000  # Aylık gelir

# Kredi almak için HEM 18 yaş HEM DE 3000 TL gelir gerekli
if yas >= 18 and gelir >= 3000:
    print("Kredi alabilirsiniz")      # Her iki koşul da True
else:
    print("Kredi alamazsınız")        # En az biri False
````

````python
# or operatörü - En az bir koşul True olmalı
# Koşullardan herhangi biri sağlansa yeterli
ogrenci = True   # Öğrenci mi?
emekli = False   # Emekli mi?

# İndirim için öğrenci VEYA emekli olmak yeterli
if ogrenci or emekli:
    print("İndirimli bilet alabilirsiniz")  # En az biri True
else:
    print("Normal fiyat uygulanır")          # İkisi de False
````

````python
# not operatörü - Koşulun tersini alır
# True → False, False → True
giris_yapildi = False  # Kullanıcı giriş yapmış mı?

# not False → True olur
if not giris_yapildi:
    print("Lütfen giriş yapınız")  # giris_yapildi False olduğu için çalışır

# Aynı mantık şöyle de yazılabilir:
# if giris_yapildi == False:
````

````python
# Mantıksal operatörlerin birlikte kullanımı
# Parantezlerle öncelik belirlenebilir
yas = 22
gelir = 4000
ogrenci = True

# Karmaşık koşul: (18 yaş VE 5000 gelir) VEYA (öğrenci VE 18 yaş)
if (yas >= 18 and gelir >= 5000) or (ogrenci and yas >= 18):
    print("Özel kampanyaya dahilsiniz")
else:
    print("Üzgünüz, kampanya koşullarını sağlamıyorsunuz")
````

# **3.6 İç İçe Koşullar (Nested Conditions)**

Bir koşul bloğunun içinde başka koşul blokları tanımlanabilir. Bu yapıya **iç içe koşullar** denir.

```python
if dış_koşul:
    # dış koşul True
    if iç_koşul:
        # iç koşul da True
    else:
        # iç koşul False
else:
    # dış koşul False
```

**Dikkat:** İç içe koşullarda girintileme çok önemlidir. Her iç blok 4 boşluk daha girintili olmalıdır.

````python
# İç içe koşullar örneği - Karar ağacı
# Dış koşul sağlanırsa iç koşul kontrol edilir
isgunu = input("Bugün iş günü mü? (evet/hayır): ")

if isgunu == "evet":  # Dış koşul: İş günü mü?
    print("İşe gitmek için hazırlan")
    
    # İç koşul: Sadece iş günü ise trafik sorulur
    trafik = input("Trafik var mı? (var/yok): ")
    
    if trafik == "var":  # İç koşul kontrolü
        print("Toplu taşıma kullan")
    else:
        print("Araba ile gidebilirsin")
else:  # Dış koşul False: İş günü değil
    print("Uyumaya devam edebilirsin")
````

# **3.7 Pratik Örnekler**

````python
# Pratik Örnek 1: Prim Hesaplama
# Ciroya göre farklı prim oranları uygulama

ciro = int(input("Ciro miktarını girin (TL): "))

# Geçerli aralık kontrolü
if ciro < 100 or ciro > 2000:
    print("Lütfen 100 ile 2000 arasında değer girin")  # Geçersiz değer
elif ciro >= 100 and ciro <= 499:   # 100-499 TL arası
    prim = ciro * 0.10              # %10 prim
    print(f"Prim (%10): {prim:.2f} TL")
elif ciro >= 500 and ciro <= 999:   # 500-999 TL arası
    prim = ciro * 0.20              # %20 prim
    print(f"Prim (%20): {prim:.2f} TL")
else:                                # 1000-2000 TL arası
    prim = ciro * 0.30              # %30 prim
    print(f"Prim (%30): {prim:.2f} TL")
````

````python
# Pratik Örnek 2: En Büyük Sayıyı Bulma
# Üç sayı arasından en büyüğünü belirleme

a = int(input("1. sayıyı girin: "))
b = int(input("2. sayıyı girin: "))
c = int(input("3. sayıyı girin: "))

# Her sayıyı diğer ikisiyle karşılaştır
if a > b and a > c:           # a hem b'den hem c'den büyük mü?
    print(f"En büyük sayı: {a}")
elif b > a and b > c:         # b hem a'dan hem c'den büyük mü?
    print(f"En büyük sayı: {b}")
else:                          # c en büyük (veya eşit)
    print(f"En büyük sayı: {c}")
````

````python
# Pratik Örnek 3: Meyve Tanıma
# if-elif-else zinciri ile string eşleştirme

soru = input("Bir meyve adı söyleyin: ")

# Bilinen meyveleri kontrol et
if soru == "elma":
    print("Evet, elma bir meyvedir.")
elif soru == "karpuz":
    print("Evet, karpuz bir meyvedir.")
elif soru == "armut":
    print("Evet, armut bir meyvedir.")
elif soru == "çilek":
    print("Evet, çilek bir meyvedir.")
else:
    # Bilinmeyen bir şey girildi
    print(f"{soru} gerçekten bir meyve midir?")
````
