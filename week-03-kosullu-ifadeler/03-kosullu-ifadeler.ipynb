# **3. Koşullu İfadeler**

Bu derste Python'da koşullu ifadeleri (if-elif-else) ve karşılaştırma/mantıksal operatörleri öğreneceğiz.

## **3.1 if-elif-else Yapısı**

Koşullu ifadeler, programın belirli koşullara göre farklı kod bloklarını çalıştırmasını sağlar.

```python
# Basit if örneği
yas = 20

if yas >= 18:
    print("Reşitsiniz")
```

```python
# if-else örneği
yas = 15

if yas >= 18:
    print("Reşitsiniz")
else:
    print("Reşit değilsiniz")
```

```python
# if-elif-else örneği: Sınav puanı değerlendirme
puan = int(input("Sınav puanınızı girin: "))

if puan >= 0 and puan <= 49:
    print("Kötü")
elif puan >= 50 and puan <= 79:
    print("İyi")
elif puan >= 80 and puan <= 100:
    print("Çok iyi")
else:
    print("Lütfen 0 ile 100 arasında puan giriniz")
```

## **3.2 Karşılaştırma Operatörleri**

```python
a = 10
b = 5

print(f"{a} == {b}:", a == b)  # Eşit mi?
print(f"{a} != {b}:", a != b)  # Eşit değil mi?
print(f"{a} > {b}:", a > b)    # Büyük mü?
print(f"{a} < {b}:", a < b)    # Küçük mü?
print(f"{a} >= {b}:", a >= b)  # Büyük eşit mi?
print(f"{a} <= {b}:", a <= b)  # Küçük eşit mi?
```

## **3.3 Mantıksal Operatörler**

```python
# and operatörü - Her iki koşul da doğru olmalı
yas = 25
gelir = 5000

if yas >= 18 and gelir >= 3000:
    print("Kredi alabilirsiniz")
else:
    print("Kredi alamazsınız")
```

```python
# or operatörü - En az bir koşul doğru olmalı
ogrenci = True
emekli = False

if ogrenci or emekli:
    print("İndirimli bilet alabilirsiniz")
```

```python
# not operatörü - Koşulu tersine çevirir
giris_yapildi = False

if not giris_yapildi:
    print("Lütfen giriş yapınız")
```

## **3.4 Pratik Örnek: Prim Hesaplama**

```python
ciro = int(input("Ciro miktarını girin: "))

if ciro < 100 or ciro > 2000:
    print("Ciro değerini 100 ile 2000 arasında girin")
elif ciro >= 100 and ciro <= 499:
    prim = ciro * 0.1
    print(f"Prim (%10): {prim} TL")
elif ciro >= 500 and ciro <= 999:
    prim = ciro * 0.2
    print(f"Prim (%20): {prim} TL")
elif ciro >= 1000 and ciro <= 2000:
    prim = ciro * 0.3
    print(f"Prim (%30): {prim} TL")
```

## **3.5 Pratik Örnek: En Büyük Sayıyı Bulma**

```python
a = int(input("1. sayıyı girin: "))
b = int(input("2. sayıyı girin: "))
c = int(input("3. sayıyı girin: "))

if a > b and a > c:
    print(f"En büyük sayı: {a}")
elif b > a and b > c:
    print(f"En büyük sayı: {b}")
else:
    print(f"En büyük sayı: {c}")
```

## **3.6 İç İçe Koşullar**

```python
isgunu = input("Bugün iş günü mü? (evet/hayır): ")

if isgunu == "evet":
    print("İşe gitmek için hazırlan")
    trafik = input("Trafik var mı? (var/yok): ")
    if trafik == "var":
        print("Vapur ile git")
    else:
        print("Araba ile git")
else:
    print("Uyumaya devam et")
```
