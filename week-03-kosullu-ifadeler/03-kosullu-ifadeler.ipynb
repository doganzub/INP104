### if Deyiminin SÃ¶z Dizimi

```python

    # koÅŸul doÄŸruysa bu blok Ã§alÄ±ÅŸÄ±r
```

**Ã–nemli Kurallar:**
- `if` satÄ±rÄ±nÄ±n sonunda `:` (iki nokta) olmalÄ±dÄ±r
- KoÅŸul bloÄŸu **4 boÅŸluk** girintili yazÄ±lmalÄ±dÄ±r

````python
# if deyimi - Basit koÅŸul kontrolÃ¼
# KoÅŸul doÄŸruysa (True) blok iÃ§indeki kod Ã§alÄ±ÅŸÄ±r
# KoÅŸul yanlÄ±ÅŸsa (False) blok atlanÄ±r
n = 255  # Kontrol edilecek deÄŸiÅŸken

# n'nin 10'dan bÃ¼yÃ¼k olup olmadÄ±ÄŸÄ±nÄ± kontrol et
# > iÅŸareti "bÃ¼yÃ¼ktÃ¼r" anlamÄ±na gelir
if n > 10:
    print("SayÄ± 10'dan bÃ¼yÃ¼ktÃ¼r!")  # Bu satÄ±r girintili yazÄ±lmalÄ± (4 boÅŸluk)
````

# **3.2 KarÅŸÄ±laÅŸtÄ±rma OperatÃ¶rleri**

KoÅŸullu ifadelerde iki deÄŸeri karÅŸÄ±laÅŸtÄ±rmak iÃ§in **karÅŸÄ±laÅŸtÄ±rma operatÃ¶rleri** kullanÄ±lÄ±r. Bu operatÃ¶rler her zaman `True` veya `False` (Boolean) deÄŸer dÃ¶ndÃ¼rÃ¼r.

| OperatÃ¶r | AnlamÄ± | Ã–rnek | SonuÃ§ |
|----------|--------|-------|-------|
| `>` | BÃ¼yÃ¼ktÃ¼r | `10 > 5` | `True` |
| `<` | KÃ¼Ã§Ã¼ktÃ¼r | `10 < 5` | `False` |
| `>=` | BÃ¼yÃ¼k eÅŸittir | `10 >= 10` | `True` |
| `<=` | KÃ¼Ã§Ã¼k eÅŸittir | `5 <= 10` | `True` |
| `==` | EÅŸittir | `5 == 5` | `True` |
| `!=` | EÅŸit deÄŸildir | `5 != 10` | `True` |

**Dikkat:** `=` (tek eÅŸittir) **atama** operatÃ¶rÃ¼dÃ¼r, `==` (Ã§ift eÅŸittir) **karÅŸÄ±laÅŸtÄ±rma** operatÃ¶rÃ¼dÃ¼r!

````python
# KarÅŸÄ±laÅŸtÄ±rma operatÃ¶rleri - Her biri True veya False dÃ¶ndÃ¼rÃ¼r
a = 10  # Ä°lk deÄŸer
b = 5   # Ä°kinci deÄŸer

# == operatÃ¶rÃ¼: Ä°ki deÄŸer eÅŸit mi?
print(f"{a} == {b}:", a == b)   # False - 10, 5'e eÅŸit deÄŸil

# != operatÃ¶rÃ¼: Ä°ki deÄŸer farklÄ± mÄ±?
print(f"{a} != {b}:", a != b)   # True - 10, 5'ten farklÄ±

# > operatÃ¶rÃ¼: Sol taraf saÄŸdan bÃ¼yÃ¼k mÃ¼?
print(f"{a} > {b}:", a > b)     # True - 10, 5'ten bÃ¼yÃ¼k

# < operatÃ¶rÃ¼: Sol taraf saÄŸdan kÃ¼Ã§Ã¼k mÃ¼?
print(f"{a} < {b}:", a < b)     # False - 10, 5'ten kÃ¼Ã§Ã¼k deÄŸil

# >= operatÃ¶rÃ¼: Sol taraf saÄŸdan bÃ¼yÃ¼k veya eÅŸit mi?
print(f"{a} >= {b}:", a >= b)   # True - 10, 5'ten bÃ¼yÃ¼k

# <= operatÃ¶rÃ¼: Sol taraf saÄŸdan kÃ¼Ã§Ã¼k veya eÅŸit mi?
print(f"{a} <= {b}:", a <= b)   # False - 10, 5'e kÃ¼Ã§Ã¼k veya eÅŸit deÄŸil
````

````python
# = ve == arasÄ±ndaki fark - Ã‡OK Ã–NEMLÄ°!
# = â†’ DeÄŸer atama (assignment)
# == â†’ EÅŸitlik karÅŸÄ±laÅŸtÄ±rmasÄ± (comparison)

x = 26       # x deÄŸiÅŸkenine 26 deÄŸerini ATA
print("x =", x)  # x = 26

# x'in 26'ya eÅŸit olup olmadÄ±ÄŸÄ±nÄ± KONTROL ET
sonuc = x == 26  # SonuÃ§ True veya False olacak
print("x == 26:", sonuc)  # True
````

# **3.3 if-else YapÄ±sÄ±**

`else` deyimi, `if` koÅŸulu saÄŸlanmadÄ±ÄŸÄ±nda Ã§alÄ±ÅŸacak alternatif bir blok tanÄ±mlar.

```python
if koÅŸul:
    # koÅŸul True ise bu blok Ã§alÄ±ÅŸÄ±r
else:
    # koÅŸul False ise bu blok Ã§alÄ±ÅŸÄ±r
```

`else` deyiminin anlamÄ±: **"EÄŸer yukarÄ±daki koÅŸul(lar) gerÃ§ekleÅŸmezse..."**

````python
# if-else yapÄ±sÄ± - Ä°ki alternatif durum
# KoÅŸul doÄŸruysa if bloÄŸu, yanlÄ±ÅŸsa else bloÄŸu Ã§alÄ±ÅŸÄ±r
yas = 15  # Kontrol edilecek yaÅŸ deÄŸeri

if yas >= 18:  # YaÅŸ 18 veya Ã¼stÃ¼ mÃ¼?
    print("ReÅŸitsiniz")  # Bu blok yas >= 18 olduÄŸunda Ã§alÄ±ÅŸÄ±r
else:
    print("ReÅŸit deÄŸilsiniz")  # Bu blok yas < 18 olduÄŸunda Ã§alÄ±ÅŸÄ±r
````

````python
# Parola kontrolÃ¼ Ã¶rneÄŸi
# if-else yapÄ±sÄ±yla basit bir giriÅŸ sistemi
parola = input("Parola: ")  # KullanÄ±cÄ±dan parola al

# Girilen parola doÄŸru mu kontrol et
if parola == "12345678":  # == ile karÅŸÄ±laÅŸtÄ±rma
    print("Sisteme HoÅŸgeldiniz!")  # DoÄŸru parola
else:
    print("YanlÄ±ÅŸ parola girdiniz!")  # YanlÄ±ÅŸ parola
````

# **3.4 if-elif-else YapÄ±sÄ±**

`elif` (else if), birden fazla koÅŸulun sÄ±rayla kontrol edilmesi gerektiÄŸinde kullanÄ±lÄ±r.

```python
if koÅŸul_1:
    # koÅŸul_1 True ise Ã§alÄ±ÅŸÄ±r
```

### if vs elif ArasÄ±ndaki Kritik Fark

| Art arda `if` bloklarÄ± | `if-elif` yapÄ±sÄ± |
|------------------------|------------------|
| Her `if` **ayrÄ± ayrÄ±** deÄŸerlendirilir | Ä°lk doÄŸru koÅŸul bulunduÄŸunda **durur** |
| TÃ¼m doÄŸru koÅŸullar Ã§alÄ±ÅŸÄ±r | YalnÄ±zca **ilk doÄŸru** koÅŸul Ã§alÄ±ÅŸÄ±r |
| Birbirinden baÄŸÄ±msÄ±z kontroller | Birbirine baÄŸlÄ± kontroller |

````python
# if-elif-else yapÄ±sÄ± - SÄ±nav puanÄ± deÄŸerlendirme
# Sadece ilk doÄŸru koÅŸul Ã§alÄ±ÅŸÄ±r, diÄŸerleri atlanÄ±r
puan = int(input("SÄ±nav puanÄ±nÄ±zÄ± girin: "))

# KoÅŸullar sÄ±rayla kontrol edilir
if puan >= 0 and puan <= 49:       # 0-49 arasÄ±
    print("KaldÄ±nÄ±z")              # Ä°lk koÅŸul True ise burada durur
elif puan >= 50 and puan <= 64:    # 50-64 arasÄ±
    print("GeÃ§er")                 # Ä°kinci koÅŸul True ise burada durur
elif puan >= 65 and puan <= 79:    # 65-79 arasÄ±
    print("Orta")
elif puan >= 80 and puan <= 89:    # 80-89 arasÄ±
    print("Ä°yi")
elif puan >= 90 and puan <= 100:   # 90-100 arasÄ±
    print("Pekiyi")
else:                               # HiÃ§biri deÄŸilse (geÃ§ersiz puan)
    print("LÃ¼tfen 0 ile 100 arasÄ±nda puan giriniz")
````

````python
# if vs elif farkÄ±nÄ± gÃ¶stermek iÃ§in Ã¶rnek
# AynÄ± deÄŸer (40) iÃ§in farklÄ± davranÄ±ÅŸlar

a = 40
print("=== Art arda if bloklarÄ± ===")

# Her if ayrÄ± deÄŸerlendirilir - Ä°KÄ° sonuÃ§ verir!
if a < 100:
    print("SayÄ± 100'den kÃ¼Ã§Ã¼k")  # Bu Ã§alÄ±ÅŸÄ±r (40 < 100)

if a < 50:
    print("SayÄ± 50'den kÃ¼Ã§Ã¼k")   # Bu DA Ã§alÄ±ÅŸÄ±r (40 < 50)

print("\n=== if-elif yapÄ±sÄ± ===")

# elif ile sadece Ä°LK doÄŸru koÅŸul Ã§alÄ±ÅŸÄ±r
if a < 100:
    print("SayÄ± 100'den kÃ¼Ã§Ã¼k")  # Bu Ã§alÄ±ÅŸÄ±r ve durur!
elif a < 50:
    print("SayÄ± 50'den kÃ¼Ã§Ã¼k")   # Buraya HÄ°Ã‡ ulaÅŸmaz
````

# **3.5 MantÄ±ksal OperatÃ¶rler**

Birden fazla koÅŸulu birleÅŸtirmek iÃ§in **mantÄ±ksal operatÃ¶rler** kullanÄ±lÄ±r.

| OperatÃ¶r | AnlamÄ± | AÃ§Ä±klama |
|----------|--------|----------|
| `and` | Ve | Her iki koÅŸul da `True` olmalÄ± |
| `or` | Veya | En az bir koÅŸul `True` olmalÄ± |
| `not` | DeÄŸil | KoÅŸulun tersini alÄ±r |

````python
# and operatÃ¶rÃ¼ - Her iki koÅŸul da True olmalÄ±
# TÃ¼m koÅŸullar saÄŸlanmazsa blok Ã§alÄ±ÅŸmaz
yas = 25     # KullanÄ±cÄ±nÄ±n yaÅŸÄ±
gelir = 5000  # AylÄ±k gelir

# Kredi almak iÃ§in HEM 18 yaÅŸ HEM DE 3000 TL gelir gerekli
if yas >= 18 and gelir >= 3000:
    print("Kredi alabilirsiniz")      # Her iki koÅŸul da True
else:
    print("Kredi alamazsÄ±nÄ±z")        # En az biri False
````

````python
# or operatÃ¶rÃ¼ - En az bir koÅŸul True olmalÄ±
# KoÅŸullardan herhangi biri saÄŸlansa yeterli
ogrenci = True   # Ã–ÄŸrenci mi?
emekli = False   # Emekli mi?

# Ä°ndirim iÃ§in Ã¶ÄŸrenci VEYA emekli olmak yeterli
if ogrenci or emekli:
    print("Ä°ndirimli bilet alabilirsiniz")  # En az biri True
else:
    print("Normal fiyat uygulanÄ±r")          # Ä°kisi de False
````

````python
# not operatÃ¶rÃ¼ - KoÅŸulun tersini alÄ±r
# True â†’ False, False â†’ True
giris_yapildi = False  # KullanÄ±cÄ± giriÅŸ yapmÄ±ÅŸ mÄ±?

# not False â†’ True olur
if not giris_yapildi:
    print("LÃ¼tfen giriÅŸ yapÄ±nÄ±z")  # giris_yapildi False olduÄŸu iÃ§in Ã§alÄ±ÅŸÄ±r

# AynÄ± mantÄ±k ÅŸÃ¶yle de yazÄ±labilir:
# if giris_yapildi == False:
````

````python
# MantÄ±ksal operatÃ¶rlerin birlikte kullanÄ±mÄ±
# Parantezlerle Ã¶ncelik belirlenebilir
yas = 22
gelir = 4000
ogrenci = True

# KarmaÅŸÄ±k koÅŸul: (18 yaÅŸ VE 5000 gelir) VEYA (Ã¶ÄŸrenci VE 18 yaÅŸ)
if (yas >= 18 and gelir >= 5000) or (ogrenci and yas >= 18):
    print("Ã–zel kampanyaya dahilsiniz")
else:
    print("ÃœzgÃ¼nÃ¼z, kampanya koÅŸullarÄ±nÄ± saÄŸlamÄ±yorsunuz")
````

# **3.6 Ä°Ã§ Ä°Ã§e KoÅŸullar (Nested Conditions)**

Bir koÅŸul bloÄŸunun iÃ§inde baÅŸka koÅŸul bloklarÄ± tanÄ±mlanabilir. Bu yapÄ±ya **iÃ§ iÃ§e koÅŸullar** denir.

```python

    # dÄ±ÅŸ koÅŸul True
    if iÃ§_koÅŸul:
        # iÃ§ koÅŸul da True
    else:
        # iÃ§ koÅŸul False
else:
    # dÄ±ÅŸ koÅŸul False
```

**Dikkat:** Ä°Ã§ iÃ§e koÅŸullarda girintileme Ã§ok Ã¶nemlidir. Her iÃ§ blok 4 boÅŸluk daha girintili olmalÄ±dÄ±r.

````python
# Ä°Ã§ iÃ§e koÅŸullar Ã¶rneÄŸi - Karar aÄŸacÄ±
# DÄ±ÅŸ koÅŸul saÄŸlanÄ±rsa iÃ§ koÅŸul kontrol edilir
isgunu = input("BugÃ¼n iÅŸ gÃ¼nÃ¼ mÃ¼? (evet/hayÄ±r): ")

if isgunu == "evet":  # DÄ±ÅŸ koÅŸul: Ä°ÅŸ gÃ¼nÃ¼ mÃ¼?
    print("Ä°ÅŸe gitmek iÃ§in hazÄ±rlan")
    
    # Ä°Ã§ koÅŸul: Sadece iÅŸ gÃ¼nÃ¼ ise trafik sorulur
    trafik = input("Trafik var mÄ±? (var/yok): ")
    
    if trafik == "var":  # Ä°Ã§ koÅŸul kontrolÃ¼
        print("Toplu taÅŸÄ±ma kullan")
    else:
        print("Araba ile gidebilirsin")
else:  # DÄ±ÅŸ koÅŸul False: Ä°ÅŸ gÃ¼nÃ¼ deÄŸil
    print("Uyumaya devam edebilirsin")
````

````python
# Pratik Ã–rnek 1: Prim Hesaplama
# Ciroya gÃ¶re farklÄ± prim oranlarÄ± uygulama

ciro = int(input("Ciro miktarÄ±nÄ± girin (TL): "))

# GeÃ§erli aralÄ±k kontrolÃ¼
if ciro < 100 or ciro > 2000:
    print("LÃ¼tfen 100 ile 2000 arasÄ±nda deÄŸer girin")  # GeÃ§ersiz deÄŸer
elif ciro >= 100 and ciro <= 499:   # 100-499 TL arasÄ±
    prim = ciro * 0.10              # %10 prim
    print(f"Prim (%10): {prim:.2f} TL")
elif ciro >= 500 and ciro <= 999:   # 500-999 TL arasÄ±
    prim = ciro * 0.20              # %20 prim
    print(f"Prim (%20): {prim:.2f} TL")
else:                                # 1000-2000 TL arasÄ±
    prim = ciro * 0.30              # %30 prim
    print(f"Prim (%30): {prim:.2f} TL")
````
