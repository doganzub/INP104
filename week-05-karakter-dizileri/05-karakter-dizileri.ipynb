# **5.1 Karakter Dizisi Nedir?**

Karakter dizileri (strings), Python'da **en sÄ±k kullanÄ±lan veri tiplerinden** biridir. `type()` fonksiyonu ile sorgulattÄ±ÄŸÄ±nÄ±zda `<class 'str'>` Ã§Ä±ktÄ±sÄ± veren her nesne bir karakter dizisidir.

## Ã–zellikleri

- Tek (`'...'`), Ã§ift (`"..."`) veya Ã¼Ã§ tÄ±rnak (`'''...'''` veya `"""..."""`) iÃ§inde tanÄ±mlanÄ±r
- **Immutable (deÄŸiÅŸtirilemez)** bir veri tipidir
- Her karakterin bir **indeks numarasÄ±** vardÄ±r
- Ãœzerinde **dÃ¶ngÃ¼ kurulabilir** (iterable)

## Neden Ã–nemli?

- KullanÄ±cÄ± giriÅŸleri karakter dizisi olarak gelir (`input()`)
- Dosya iÃ§erikleri karakter dizisi olarak okunur
- Web verileri genellikle karakter dizisi formatÄ±ndadÄ±r
- VeritabanÄ± kayÄ±tlarÄ±nÄ±n Ã§oÄŸu metin iÃ§erir

````python
# Karakter dizisi tanÄ±mlama yollarÄ±
# Python'da tek, Ã§ift veya Ã¼Ã§ tÄ±rnak kullanÄ±labilir

tek_tirnak = 'Python'           # Tek tÄ±rnak ile
cift_tirnak = "Python"          # Ã‡ift tÄ±rnak ile
uc_tirnak = '''Python'''        # ÃœÃ§ tÄ±rnak ile
cok_satirli = """Python
Programlama
Dili"""                          # Ã‡ok satÄ±rlÄ± metin

# Hepsinin tipi 'str' (string)
print(type(tek_tirnak))  # <class 'str'>
print(type(cok_satirli)) # <class 'str'>
````

# **5.2 Ä°ndeksleme (Indexing)**

Python'da karakter dizilerinin her bir karakterine bir **indeks numarasÄ±** ile eriÅŸilebilir.

## Ã–nemli Kurallar

- Python'da indeksler **0'dan baÅŸlar**
- Son karaktere ulaÅŸmak iÃ§in **negatif indeks** kullanÄ±labilir
- Karakter dizisinin uzunluÄŸunu aÅŸan indeks **IndexError** verir

## Ä°ndeks Tablosu

```
P  y  t  h  o  n
0  1  2  3  4  5   â† Pozitif indeks
-6 -5 -4 -3 -2 -1  â† Negatif indeks
```

````python
# Ä°ndeksleme - KÃ¶ÅŸeli parantez ile karakter eriÅŸimi
# FormÃ¼l: karakter_dizisi[indeks_numarasi]

kelime = "Python"  # 6 karakterli bir karakter dizisi

# Pozitif indeksler (soldan saÄŸa, 0'dan baÅŸlar)
print(kelime[0])   # P - ilk karakter (indeks 0)
print(kelime[1])   # y - ikinci karakter (indeks 1)
print(kelime[5])   # n - altÄ±ncÄ± karakter (indeks 5)

# Negatif indeksler (saÄŸdan sola, -1'den baÅŸlar)
print(kelime[-1])  # n - son karakter
print(kelime[-2])  # o - sondan ikinci karakter
print(kelime[-6])  # P - sondan altÄ±ncÄ± karakter = ilk karakter
````

````python
# Ä°ndeksleme - len() fonksiyonu ile son karakter
# Son karakterin indeksi = len(karakter_dizisi) - 1

kelime = "istanbul"  # 8 karakterli

# len() fonksiyonu karakter sayÄ±sÄ±nÄ± verir
uzunluk = len(kelime)  # 8
print(f"Kelime uzunluÄŸu: {uzunluk}")

# Son karakterin indeksi = uzunluk - 1
# Ã‡Ã¼nkÃ¼ indeksler 0'dan baÅŸlar
son_indeks = len(kelime) - 1  # 7
print(f"Son karakter: {kelime[son_indeks]}")  # l

# AynÄ± iÅŸlem negatif indeks ile Ã§ok daha kolay
print(f"Son karakter: {kelime[-1]}")  # l
````

````python
# Ä°ndeksleme - Her karaktere dÃ¶ngÃ¼ ile eriÅŸim
# range(len(kelime)) ile indeks numaralarÄ±nÄ± elde et

isim = input("Ä°sminizi girin: ")

# range(len(isim)) â†’ 0'dan isim uzunluÄŸuna kadar sayÄ±lar Ã¼retir
for i in range(len(isim)):
    # i deÄŸiÅŸkeni sÄ±rayla 0, 1, 2, ... deÄŸerlerini alÄ±r
    # isim[i] ile o indeksteki karaktere ulaÅŸÄ±rÄ±z
    print(f"Ä°sminizin {i+1}. harfi: {isim[i]}")
    # i+1 kullandÄ±k Ã§Ã¼nkÃ¼ insanlar 1'den saymaya baÅŸlar
````

# **5.3 Dilimleme (Slicing)**

Dilimleme, bir karakter dizisinden **belirli bir bÃ¶lÃ¼mÃ¼ ayÄ±rmak** iÃ§in kullanÄ±lÄ±r.

## SÃ¶z Dizimi

```python
karakter_dizisi[baÅŸlangÄ±Ã§:bitiÅŸ:adÄ±m]
```

**Ã–nemli Kurallar:**
- `baÅŸlangÄ±Ã§` indeksi **dahildir**
- `bitiÅŸ` indeksi **dahil deÄŸildir**
- `adÄ±m` belirtilmezse varsayÄ±lan olarak 1'dir
- Belirtilmeyen deÄŸerler iÃ§in varsayÄ±lanlar kullanÄ±lÄ±r

````python
# Dilimleme - Temel Ã¶rnekler
# [baÅŸlangÄ±Ã§:bitiÅŸ] â†’ baÅŸlangÄ±Ã§ dahil, bitiÅŸ dahil deÄŸil

site = "www.istihza.com"

# [4:11] â†’ 4. indeksten 10. indekse kadar (11 dahil deÄŸil)
print(site[4:11])   # istihza

# [12:15] â†’ 12. indeksten 14. indekse kadar
print(site[12:15])  # com

# [0:3] â†’ 0. indeksten 2. indekse kadar
print(site[0:3])    # www
````

````python
# Dilimleme - KÄ±sa yazÄ±m
# BaÅŸlangÄ±Ã§ 0 ise yazÄ±lmayabilir
# BitiÅŸ son karakter ise yazÄ±lmayabilir

metin = "Python Programlama"

# [:6] â†’ BaÅŸtan 6. indekse kadar (0:6 ile aynÄ±)
print(metin[:6])    # Python

# [7:] â†’ 7. indeksten sona kadar
print(metin[7:])    # Programlama

# [:] â†’ TamamÄ±nÄ± al (kopya)
print(metin[:])     # Python Programlama
````

````python
# Dilimleme - Negatif indekslerle
# Negatif sayÄ±lar sondan saymayÄ± saÄŸlar

parola = "gizli12345"

# [:-5] â†’ BaÅŸtan baÅŸla, son 5 karakter hariÃ§
print(parola[:-5])   # gizli

# [-5:] â†’ Son 5 karakteri al
print(parola[-5:])   # 12345

# [4:-4] â†’ 4. indeksten baÅŸla, sondan 4 karakter hariÃ§ sonuna kadar
url = "www.google.com"
print(url[4:-4])     # google
````

````python
# Dilimleme - AdÄ±m deÄŸeri ile
# [baÅŸlangÄ±Ã§:bitiÅŸ:adÄ±m] â†’ adÄ±m kadar atlayarak al

sayilar = "0123456789"

# [::2] â†’ TamamÄ±nÄ± 2'ÅŸer atlayarak al
print(sayilar[::2])   # 02468 (Ã§ift indeksler)

# [1::2] â†’ 1. indeksten baÅŸlayÄ±p 2'ÅŸer atlayarak al
print(sayilar[1::2])  # 13579 (tek indeksler)

# [::3] â†’ TamamÄ±nÄ± 3'er atlayarak al
print(sayilar[::3])   # 0369
````

````python
# Dilimleme - Ters Ã§evirme (reverse)
# [::-1] â†’ TamamÄ±nÄ± tersten al

kelime = "Python"

# [::-1] â†’ Sondan baÅŸa, tÃ¼m karakterleri al
print(kelime[::-1])   # nohtyP

# Palindrom kontrolÃ¼
test = "kayak"
if test == test[::-1]:  # Tersi kendisine eÅŸitse
    print(f"'{test}' bir palindromdur")
else:
    print(f"'{test}' palindrom deÄŸildir")
````

# **5.4 Karakter Dizileri DeÄŸiÅŸtirilemez (Immutable)**

Python'da karakter dizileri **immutable** (deÄŸiÅŸtirilemez) bir veri tipidir. Bir kez oluÅŸturulduktan sonra **iÃ§erikleri doÄŸrudan deÄŸiÅŸtirilemez**.

```python
# HATALI - DoÄŸrudan deÄŸiÅŸiklik yapÄ±lamaz!
kelime = "python"
kelime[0] = "P"  # TypeError: 'str' object does not support item assignment
```

**Ã‡Ã¶zÃ¼m:** DeÄŸiÅŸiklik yapmak iÃ§in karakter dizisini **yeniden tanÄ±mlayÄ±n**.

````python
# Karakter dizisi deÄŸiÅŸtirme - Dilimleme ile yeniden oluÅŸturma
# Karakter dizisinin bir kÄ±smÄ±nÄ± deÄŸiÅŸtirmek iÃ§in dilimlemeyi kullan

meyve = "elma"

# Ä°lk harfi bÃ¼yÃ¼tmek iÃ§in:
# "E" + meyve[1:] â†’ "E" + "lma" â†’ "Elma"
meyve = "E" + meyve[1:]
print(meyve)  # Elma

# Ortadaki bir kÄ±smÄ± deÄŸiÅŸtirmek iÃ§in:
kardiz = "istihza"
# kardiz[:3] â†’ "ist"
# kardiz[5:] â†’ "za"
kardiz = kardiz[:3] + "Ä°H" + kardiz[5:]
print(kardiz)  # istÄ°Hza
````

````python
# Karakter dizisi deÄŸiÅŸtirme - id() ile hafÄ±za adresi kontrolÃ¼
# Her yeniden atama yeni bir nesne oluÅŸturur

kardiz = "istihza"  # Ä°lk karakter dizisi
print(f"Ä°lk id: {id(kardiz)}")  # Bellek adresi

kardiz = "Ä°" + kardiz[1:]  # Yeniden tanÄ±mlama
print(f"Yeni id: {id(kardiz)}")  # FarklÄ± bellek adresi!

# id() deÄŸerleri farklÄ± â†’ FarklÄ± nesneler
# Eski nesne silindi, yeni nesne oluÅŸturuldu
````

## 5.5.1 replace() - DeÄŸiÅŸtirme

Karakter dizisi iÃ§indeki karakterleri baÅŸka karakterlerle deÄŸiÅŸtirir.

```python
karakter_dizisi.replace(eski, yeni, adet)
```

````python
# replace() metodu - Karakter/metin deÄŸiÅŸtirme
# Orijinal karakter dizisi deÄŸiÅŸmez, yeni bir kopya dÃ¶ner

kelime = "memleket"

# TÃ¼m "e" harflerini "E" ile deÄŸiÅŸtir
yeni = kelime.replace("e", "E")
print(yeni)  # mEmlEkEt

# TÃ¼m "e" harflerini sil (boÅŸ string ile deÄŸiÅŸtir)
yeni = kelime.replace("e", "")
print(yeni)  # mmlkt

# Sadece ilk 2 "e" harfini deÄŸiÅŸtir (Ã¼Ã§Ã¼ncÃ¼ parametre)
yeni = kelime.replace("e", "E", 2)
print(yeni)  # mEmlEket

# Orijinal deÄŸiÅŸmedi!
print(kelime)  # memleket
````

````python
# replace() metodu - Pratik Ã¶rnek
# URL'lerdeki "www." kÄ±smÄ±nÄ± kaldÄ±rma

site1 = "www.google.com"
site2 = "www.istihza.com"
site3 = "www.python.org"

# Her siteden "www." kÄ±smÄ±nÄ± kaldÄ±r
for site in [site1, site2, site3]:
    # replace("www.", "") â†’ "www." yerine hiÃ§bir ÅŸey koy = sil
    temiz = site.replace("www.", "")
    print(temiz)  # google.com, istihza.com, python.org
````

## 5.5.2 split() ve join() - BÃ¶lme ve BirleÅŸtirme

`split()` karakter dizisini parÃ§alara ayÄ±rÄ±r, `join()` parÃ§alarÄ± birleÅŸtirir.

````python
# split() metodu - Karakter dizisini bÃ¶lme
# VarsayÄ±lan olarak boÅŸluk karakterinden bÃ¶ler

cumle = "Ä°stanbul BÃ¼yÃ¼kÅŸehir Belediyesi"

# BoÅŸluklardan bÃ¶l
parcalar = cumle.split()
print(parcalar)  # ['Ä°stanbul', 'BÃ¼yÃ¼kÅŸehir', 'Belediyesi']

# For dÃ¶ngÃ¼sÃ¼ ile her parÃ§aya eriÅŸ
for kelime in cumle.split():
    print(kelime)
````

````python
# split() metodu - Belirli karakterden bÃ¶lme
# Parametre olarak ayÄ±rÄ±cÄ± karakter verilebilir

sehirler = "Bolvadin, Kilis, Siverek, Ä°skenderun, Ä°stanbul"

# VirgÃ¼lden bÃ¶l
liste = sehirler.split(",")
print(liste)
# [' Bolvadin', ' Kilis', ' Siverek', ' Ä°skenderun', ' Ä°stanbul']

# ", " (virgÃ¼l ve boÅŸluk) karakterinden bÃ¶l - daha temiz
liste = sehirler.split(", ")
print(liste)
# ['Bolvadin', 'Kilis', 'Siverek', 'Ä°skenderun', 'Ä°stanbul']
````

````python
# split() metodu - Pratik Ã¶rnek: KÄ±saltma Ã§Ä±karma
# Kurum adÄ±nÄ±n baÅŸ harflerini al

kurum = input("Kurum adÄ±nÄ± girin: ")

kisaltma = ""  # Sonucu biriktireceÄŸimiz deÄŸiÅŸken

# Kurum adÄ±nÄ± kelimelerine bÃ¶l
for kelime in kurum.split():
    # Her kelimenin ilk harfini ekle (kelime[0])
    kisaltma += kelime[0]

print(f"KÄ±saltma: {kisaltma.upper()}")  # BÃ¼yÃ¼k harfe Ã§evir
````

````python
# join() metodu - Liste elemanlarÄ±nÄ± birleÅŸtirme
# Syntax: "ayÄ±rÄ±cÄ±".join(liste)

harfler = ["P", "y", "t", "h", "o", "n"]

# BoÅŸluksuz birleÅŸtir
kelime = "".join(harfler)
print(kelime)  # Python

# Tire ile birleÅŸtir
kelime = "-".join(harfler)
print(kelime)  # P-y-t-h-o-n

# BoÅŸluk ile birleÅŸtir
kelimeler = ["Python", "Ã¶ÄŸreniyorum"]
cumle = " ".join(kelimeler)
print(cumle)  # Python Ã¶ÄŸreniyorum
````

## 5.5.3 lower() ve upper() - Harf DÃ¶nÃ¼ÅŸÃ¼mÃ¼

````python
# lower() ve upper() metotlarÄ±
# KÃ¼Ã§Ã¼k/bÃ¼yÃ¼k harfe dÃ¶nÃ¼ÅŸtÃ¼rme

metin = "Python Programlama"

# TamamÄ±nÄ± kÃ¼Ã§Ã¼k harfe Ã§evir
print(metin.lower())  # python programlama

# TamamÄ±nÄ± bÃ¼yÃ¼k harfe Ã§evir
print(metin.upper())  # PYTHON PROGRAMLAMA

# Orijinal metin deÄŸiÅŸmez!
print(metin)  # Python Programlama
````

````python
# lower/upper - TÃ¼rkÃ§e karakter sorunu
# Python'un lower() metodu "I" â†’ "Ä±" dÃ¶nÃ¼ÅŸÃ¼mÃ¼nÃ¼ yapmaz!

metin = "ISIK Ä°SÄ°M"

# Standart lower() TÃ¼rkÃ§e iÃ§in hatalÄ± Ã§alÄ±ÅŸÄ±r
print(metin.lower())  # isik isim (yanlÄ±ÅŸ: Ä±ÅŸÄ±k olmalÄ±)

# Ã‡Ã¶zÃ¼m: Ã–nce TÃ¼rkÃ§e karakterleri manuel deÄŸiÅŸtir
turkce = metin.replace("I", "Ä±").replace("Ä°", "i")
print(turkce.lower())  # Ä±ÅŸÄ±k isim (doÄŸru)
````

````python
# capitalize(), title(), swapcase()
# DiÄŸer harf dÃ¶nÃ¼ÅŸÃ¼m metotlarÄ±

metin = "python programlama DÄ°LÄ°"

# capitalize() - Sadece ilk harfi bÃ¼yÃ¼k yapar
print(metin.capitalize())  # Python programlama dili

# title() - Her kelimenin ilk harfini bÃ¼yÃ¼k yapar
print(metin.title())  # Python Programlama Dili

# swapcase() - BÃ¼yÃ¼k/kÃ¼Ã§Ã¼k harfleri yer deÄŸiÅŸtirir
print(metin.swapcase())  # PYTHON PROGRAMLAMA dÄ°lÄ°
````

## 5.5.4 count() ve find() - Arama MetotlarÄ±

````python
# count() metodu - Karakter/metin sayma
# Belirtilen deÄŸerin kaÃ§ kez geÃ§tiÄŸini dÃ¶ndÃ¼rÃ¼r

sehir = "adana"

# "a" harfi kaÃ§ kez geÃ§iyor?
print(sehir.count("a"))  # 3

# "n" harfi kaÃ§ kez geÃ§iyor?
print(sehir.count("n"))  # 1

# "an" ifadesi kaÃ§ kez geÃ§iyor?
print(sehir.count("an"))  # 1

# Olmayan karakter
print(sehir.count("z"))  # 0
````

````python
# find() metodu - Karakter/metin konumu bulma
# Ä°lk bulunan konumun indeksini dÃ¶ndÃ¼rÃ¼r
# Bulunamazsa -1 dÃ¶ndÃ¼rÃ¼r

metin = "Python programlama"

# "prog" ifadesinin baÅŸlangÄ±Ã§ indeksi
print(metin.find("prog"))  # 7

# "th" ifadesinin baÅŸlangÄ±Ã§ indeksi
print(metin.find("th"))  # 2

# Bulunamayan deÄŸer
print(metin.find("Java"))  # -1

# Ä°kinci parametre: Aramaya baÅŸlanacak indeks
sehir = "adana"
print(sehir.find("a"))     # 0 - ilk "a"
print(sehir.find("a", 1))  # 2 - 1. indeksten sonraki ilk "a"
````

````python
# index() metodu - find() gibi ama hata verir
# Bulunamazsa ValueError hatasÄ± fÄ±rlatÄ±r

metin = "Python"

# "th" ifadesini bul
print(metin.index("th"))  # 2

# Bulunamayan deÄŸer - HATA!
# print(metin.index("Java"))  # ValueError: substring not found

# find() daha gÃ¼venlidir Ã§Ã¼nkÃ¼ -1 dÃ¶ndÃ¼rÃ¼r, hata vermez
print(metin.find("Java"))  # -1
````

## 5.5.5 strip(), lstrip(), rstrip() - BoÅŸluk Temizleme

````python
# strip() metotlarÄ± - BaÅŸtan/sondan karakter silme
# VarsayÄ±lan olarak boÅŸluk karakterlerini siler

metin = "   Python   "

# strip() - Hem baÅŸtan hem sondan temizle
print(f"[{metin.strip()}]")   # [Python]

# lstrip() - Sadece soldan (left) temizle
print(f"[{metin.lstrip()}]")  # [Python   ]

# rstrip() - Sadece saÄŸdan (right) temizle
print(f"[{metin.rstrip()}]")  # [   Python]

# Belirli karakterleri silme
url = "www.google.com"
print(url.strip("wcom."))  # google (baÅŸtan ve sondan siler)
````

## 5.5.6 startswith() ve endswith() - BaÅŸlangÄ±Ã§/BitiÅŸ KontrolÃ¼

````python
# startswith() ve endswith() metotlarÄ±
# Karakter dizisinin baÅŸlangÄ±Ã§/bitiÅŸini kontrol eder
# True veya False dÃ¶ndÃ¼rÃ¼r

dosya = "rapor.pdf"

# Dosya adÄ± ".pdf" ile bitiyor mu?
print(dosya.endswith(".pdf"))    # True
print(dosya.endswith(".txt"))    # False

# Dosya adÄ± "rapor" ile baÅŸlÄ±yor mu?
print(dosya.startswith("rapor")) # True
print(dosya.startswith("Ã¶zet"))  # False

# Birden fazla seÃ§enek kontrolÃ¼ (tuple ile)
print(dosya.endswith((".pdf", ".doc", ".txt")))  # True
````

## 5.5.7 isXXX() - Karakter Tipi Kontrol MetotlarÄ±

````python
# isXXX() metotlarÄ± - Karakter tipi kontrolÃ¼
# TÃ¼m karakterler belirtilen tipten mi? True/False dÃ¶ner

# isdigit() - Sadece rakamlardan mÄ± oluÅŸuyor?
print("12345".isdigit())    # True
print("12.5".isdigit())     # False (nokta var)
print("12a".isdigit())      # False (harf var)

# isalpha() - Sadece harflerden mi oluÅŸuyor?
print("Python".isalpha())   # True
print("Python3".isalpha())  # False (sayÄ± var)

# isalnum() - Harf VEYA rakamdan mÄ± oluÅŸuyor?
print("Python3".isalnum())  # True
print("Python 3".isalnum()) # False (boÅŸluk var)

# isspace() - Sadece boÅŸluklardan mÄ± oluÅŸuyor?
print("   ".isspace())      # True
print(" a ".isspace())      # False
````

````python
# isXXX() metotlarÄ± - BÃ¼yÃ¼k/kÃ¼Ã§Ã¼k harf kontrolÃ¼

# islower() - TÃ¼m harfler kÃ¼Ã§Ã¼k mÃ¼?
print("python".islower())   # True
print("Python".islower())   # False

# isupper() - TÃ¼m harfler bÃ¼yÃ¼k mÃ¼?
print("PYTHON".isupper())   # True
print("Python".isupper())   # False

# istitle() - BaÅŸlÄ±k formatÄ±nda mÄ±?
print("Python Programlama".istitle())  # True
print("python programlama".istitle())  # False
````

# **5.6 enumerate() Fonksiyonu**

`enumerate()` fonksiyonu, bir iterable Ã¼zerinde dÃ¶nerken **hem indeks hem de deÄŸere** aynÄ± anda eriÅŸim saÄŸlar.

````python
# enumerate() fonksiyonu - Ä°ndeks ve deÄŸer birlikte
# Her turda (indeks, deÄŸer) Ã§ifti dÃ¶ndÃ¼rÃ¼r

isim = "Python"

# Klasik yÃ¶ntem - range(len()) ile
print("=== range(len()) yÃ¶ntemi ===")
for i in range(len(isim)):
    print(f"{i}: {isim[i]}")

print()

# enumerate() ile - Daha temiz ve pythonic
print("=== enumerate() yÃ¶ntemi ===")
for indeks, harf in enumerate(isim):
    print(f"{indeks}: {harf}")
````

````python
# enumerate() fonksiyonu - BaÅŸlangÄ±Ã§ deÄŸeri
# start parametresi ile numaralandÄ±rma baÅŸlangÄ±cÄ± deÄŸiÅŸtirilebilir

meyveler = ["elma", "armut", "muz", "Ã§ilek"]

# VarsayÄ±lan (0'dan baÅŸlar)
print("=== 0'dan baÅŸlar ===")
for i, meyve in enumerate(meyveler):
    print(f"{i}. {meyve}")

print()

# 1'den baÅŸlat (daha doÄŸal numaralandÄ±rma)
print("=== 1'den baÅŸlar ===")
for no, meyve in enumerate(meyveler, start=1):
    print(f"{no}. {meyve}")
````

````python
# Pratik Ã–rnek 1: Sesli-Sessiz Harf AyÄ±rma
# Bir kelimede sesli ve sessiz harfleri ayÄ±r

sesli_harfler = "aeÄ±ioÃ¶uÃ¼"
sessiz_harfler = "bcÃ§dfgÄŸhjklmnprsÅŸtvyz"

kelime = input("Bir kelime girin: ").lower()  # KÃ¼Ã§Ã¼k harfe Ã§evir

sesliler = ""    # Sesli harfleri biriktir
sessizler = ""   # Sessiz harfleri biriktir

for harf in kelime:
    if harf in sesli_harfler:      # Sesli harf mi?
        sesliler += harf
    elif harf in sessiz_harfler:   # Sessiz harf mi?
        sessizler += harf
    # DiÄŸer karakterler (boÅŸluk, nokta vb.) atlanÄ±r

print(f"Sesli harfler: {sesliler}")
print(f"Sessiz harfler: {sessizler}")
````

````python
# Pratik Ã–rnek 2: E-posta DoÄŸrulama
# Basit e-posta formatÄ± kontrolÃ¼

email = input("E-posta adresinizi girin: ")

# Temel kontroller
if "@" not in email:
    print("Hata: @ iÅŸareti bulunamadÄ±!")
elif email.count("@") > 1:
    print("Hata: Birden fazla @ iÅŸareti!")
elif not email.endswith((".com", ".org", ".edu", ".net", ".tr")):
    print("Hata: GeÃ§ersiz uzantÄ±!")
elif " " in email:
    print("Hata: E-posta boÅŸluk iÃ§eremez!")
else:
    # E-postayÄ± bÃ¶l
    kullanici, domain = email.split("@")
    print(f"KullanÄ±cÄ± adÄ±: {kullanici}")
    print(f"Domain: {domain}")
    print("E-posta geÃ§erli gÃ¶rÃ¼nÃ¼yor!")
````

````python
# Pratik Ã–rnek 3: Palindrom KontrolÃ¼
# Tersten okunan hali aynÄ± olan kelimeler

kelime = input("Bir kelime girin: ")

# BoÅŸluklarÄ± kaldÄ±r ve kÃ¼Ã§Ã¼k harfe Ã§evir
temiz = kelime.replace(" ", "").lower()

# Ters Ã§evir ve karÅŸÄ±laÅŸtÄ±r
ters = temiz[::-1]

if temiz == ters:
    print(f"'{kelime}' bir palindromdur!")
else:
    print(f"'{kelime}' palindrom deÄŸildir.")
    print(f"Tersi: '{ters}'")
````

````python
# Pratik Ã–rnek 4: Metin KarÅŸÄ±laÅŸtÄ±rma
# Ä°ki metin arasÄ±ndaki farklÄ±lÄ±klarÄ± bul

metin1 = "python"
metin2 = "programming"

# metin1'de olup metin2'de olmayan harfler
farkli = ""
for harf in metin1:
    if harf not in metin2:  # metin2'de yoksa
        farkli += harf

print(f"'{metin1}' iÃ§inde olup '{metin2}' iÃ§inde olmayan: {farkli}")
````
