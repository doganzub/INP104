# **5. Karakter Dizileri ve Metotları**

Bu derste Python'da karakter dizileri (string) ve string metotlarını öğreneceğiz.

## **5.1 String İndeksleme**

Python'da string karakterlerine indeks numarası ile ulaşılır. İndeksler 0'dan başlar.

```python
isim = "züber"

print(f"1. karakter: {isim[0]}")   # z
print(f"2. karakter: {isim[1]}")   # ü
print(f"Son karakter: {isim[-1]}") # r
print(f"Sondan 2.: {isim[-2]}")    # e
```

```python
# Dinamik olarak harfleri listeleme
isim = input("İsminizi girin: ")

for i in range(len(isim)):
    print(f"İsminizin {i+1}. harfi: {isim[i]}")
```

## **5.2 String Dilimleme (Slicing)**

```python
# [başlangıç:bitiş] formatı
parola = "zuberdogan1978"

print(parola[0:5])    # zuber (0-4 arası)
print(parola[5:10])   # dogan
print(parola[:5])     # zuber (baştan 5'e kadar)
print(parola[10:])    # 1978 (10'dan sona kadar)
```

```python
# Negatif indekslerle dilimleme
url = "www.zuberdogan.com"

print(url[4:-4])  # zuberdogan (baştan 4, sondan 4 hariç)
```

```python
# Ters çevirme
metin = "Python"
print(metin[::-1])  # nohtyP
```

```python
# Adım değeri ile dilimleme
sayilar = "0123456789"
print(sayilar[::2])  # 02468 (2'şer atlayarak)
```

## **5.3 replace() Metodu**

```python
# Karakter değiştirme
kelime = "memleket"

print(kelime.replace("e", ""))     # mmlkt (tüm e'leri sil)
print(kelime.replace("e", "", 1))  # mmleket (sadece ilk e'yi sil)
print(kelime.replace("m", "M", 1)) # Memleket (ilk m'yi büyüt)
```

## **5.4 lower() ve upper() Metotları**

```python
# Küçük harfe çevirme
metin = "MERHABA DÜNYA"
print(metin.lower())
```

```python
# Büyük harfe çevirme
metin = "merhaba dünya"
print(metin.upper())
```

```python
# Türkçe karakterler için özel işlem
parola = input("Parolanızı girin: ")
parola = parola.replace("I", "ı").replace("İ", "i").lower()

if parola == "test":
    print("Parola doğru")
else:
    print("Tekrar deneyin")
```

## **5.5 split() ve join() Metotları**

```python
# split() - Bölme
isim = "züber doğan"
parcalar = isim.split()
print(parcalar)  # ['züber', 'doğan']
```

```python
# Belirli karaktere göre bölme
bilgiler = "züber doğan, 1978, istanbul"
parcalar = bilgiler.split(", ")
print(parcalar)  # ['züber doğan', '1978', 'istanbul']
```

```python
# join() - Birleştirme
liste = ["züber", "doğan", "1978"]
birlesik = " ".join(liste)
print(birlesik)  # züber doğan 1978
```

## **5.6 count() ve find() Metotları**

```python
# count() - Sayma
sehir = "adana"
print(f"'a' harfi {sehir.count('a')} kez geçiyor")
```

```python
# find() / index() - Bulma
sehir = "adana"
print(f"İlk 'a' indeksi: {sehir.find('a')}")      # 0
print(f"2. 'a' indeksi: {sehir.find('a', 1)}")    # 2
```

## **5.7 enumerate() Fonksiyonu**

```python
# enumerate() ile indeks ve değer
isim = "Züber"

for indeks, harf in enumerate(isim):
    print(f"{indeks}: {harf}")
```

```python
# Başlangıç değeri ile enumerate
isim = "Züber"

for indeks, harf in enumerate(isim, start=1):
    print(f"{indeks}. harf: {harf}")
```
