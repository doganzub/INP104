# **6. Listeler ve Demetler**

Bu derste Python'da veri yapıları olan listeler ve demetleri öğreneceğiz.

## **6.1 Liste Oluşturma**

```python
# Boş liste
bos_liste = []

# Farklı veri tiplerini içeren liste
karisik_liste = ["cuma", ["züber"], 7, 5.9]
print(karisik_liste)
print(type(karisik_liste))
```

```python
# Karakter dizisinden liste oluşturma
alfabe = "abcçde"
harf_listesi = list(alfabe)
print(harf_listesi)
```

```python
# range() ile liste oluşturma
sayilar = list(range(10))
print(sayilar)
```

## **6.2 Liste Öğelerine Erişim**

```python
liste = ['Minas', 'ganimet', ['züber', 'idil', 'uygar']]

print(f"Liste uzunluğu: {len(liste)}")
print(f"1. öğe: {liste[0]}")
print(f"3. öğe (iç liste): {liste[2]}")
print(f"İç listenin 1. öğesi: {liste[2][0]}")
```

```python
# Liste öğelerini değiştirme
liste = ['Minas', 'ganimet', ['züber', 'idil', 'uygar']]

liste[0] = "Minas Doğan"
liste[1] = "Ganimet Doğan"
print(liste)
```

## **6.3 Liste Birleştirme ve Çoğaltma**

```python
# Listeye öğe ekleme (+)
liste = ['minas', 'ganimet']
liste = liste + ['orhan', 'nagehan']
print(liste)
```

```python
# İki listeyi birleştirme
zuber_aile = ['minas', 'ganimet']
idil_aile = ['orhan', 'nagehan']
tum_aile = zuber_aile + idil_aile
print(tum_aile)
```

```python
# Listeden öğe silme (del)
liste = [1, 3, 5, 7]
del liste[0]
print(liste)
```

## **6.4 Liste Metotları**

```python
# append() - Sona öğe ekle
liste = ['minas', 'ganimet']
liste.append('orhan')
print(liste)
```

```python
# insert() - Belirli konuma ekle
liste = ['can', 'ayşe']
liste.insert(1, 'sülü')  # 1. indekse ekle
print(liste)
```

```python
# remove() - Öğe sil
liste = ['can', 'sülü', 'ayşe']
liste.remove('sülü')
print(liste)
```

```python
# sort() - Sırala
sayilar = [5, 10, 76, 89, 3, 8]
sayilar.sort()
print("Artan:", sayilar)

sayilar.sort(reverse=True)
print("Azalan:", sayilar)
```

```python
# reverse() - Ters çevir
liste = ['can', 'sülü', 'ayşe']
liste.reverse()
print(liste)
```

## **6.5 Liste Kopyalama**

```python
# Referans kopyalama (YANLIŞ!)
l1 = ['can', 'murat', 'aslı']
l2 = l1  # l2, l1'e referans verir

l1[0] = 'emre'
print("l1:", l1)
print("l2:", l2)  # l2 de değişti!
```

```python
# Bağımsız kopya (DOĞRU)
l1 = ['can', 'murat', 'aslı']
l2 = list(l1)  # veya l1.copy() veya l1[:]

l1[0] = 'emre'
print("l1:", l1)
print("l2:", l2)  # l2 değişmedi
```

## **6.6 Demetler (Tuple)**

Demetler listeler gibidir ama **değiştirilemez** (immutable).

```python
# Demet oluşturma
demet = ('can', 'murat', 'aslı')
print(demet)
print(type(demet))
```

```python
# Tek öğeli demet (virgül zorunlu!)
tek_oge = ('Minas',)  # Virgüle dikkat
print(type(tek_oge))

yanlis = ('Minas')    # Bu string olur!
print(type(yanlis))
```

```python
# Demet öğelerine erişim
demet = ('Minas', 'ganimet', ('züber', 'idil', 'uygar'))

print(f"Uzunluk: {len(demet)}")
print(f"1. öğe: {demet[0]}")
print(f"İç demetin 1. öğesi: {demet[2][0]}")
```

```python
# Demetler değiştirilemez!
try:
    demet = ('a', 'b', 'c')
    demet[0] = 'x'  # Hata verir
except TypeError as hata:
    print(f"Hata: {hata}")
```
