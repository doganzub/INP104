# **8. Fonksiyonlar**

Bu derste Python'da fonksiyon tanımlama, parametreler ve return ifadesini öğreneceğiz.

## **8.1 Fonksiyon Tanımlama**

Fonksiyonlar, tekrar kullanılabilir kod bloklarıdır.

```python
# Basit fonksiyon
def selamla():
    print("Merhaba!")

# Fonksiyonu çağırma
selamla()
```

```python
# Parametreli fonksiyon
def kayit(isim, soyisim, sehir):
    print(f"İsim    : {isim}")
    print(f"Soyisim : {soyisim}")
    print(f"Şehir   : {sehir}")

# Fonksiyonu çağırma
kayit("Züber", "Doğan", "İstanbul")
```

## **8.2 Varsayılan Parametre Değerleri**

```python
# Varsayılan değerli parametreler
def selamla(isim, mesaj="Hoş geldiniz"):
    print(f"{isim}, {mesaj}!")

selamla("Züber")                    # Varsayılan mesaj
selamla("Uygar", "Günaydın")        # Özel mesaj
```

```python
# İsimli argümanlar
def islem(p1, p2, p3):
    print(f"İşlem sonucu: {p1 * p2}, Kişi: {p3}")

# İsimli argümanlarla çağırma
islem(p1=5, p2=3, p3="Züber")
islem(p3="Uygar", p1=10, p2=2)  # Sıra önemli değil
```

## **8.3 Return İfadesi**

```python
# Return ile değer döndürme
def alan_hesapla(uzunluk, genislik):
    alan = uzunluk * genislik
    return alan

sonuc = alan_hesapla(5, 3)
print(f"Alan: {sonuc} m²")
```

```python
# Çoklu değer döndürme
def dikdortgen_hesapla(uzunluk, genislik):
    alan = uzunluk * genislik
    cevre = 2 * (uzunluk + genislik)
    return alan, cevre

a, c = dikdortgen_hesapla(5, 3)
print(f"Alan: {a} m², Çevre: {c} m")
```

## **8.4 *args - Sınırsız Pozisyonel Argüman**

```python
# *args ile sınırsız argüman
def toplam(*args):
    sonuc = 0
    for sayi in args:
        sonuc += sayi
    return sonuc

print(toplam(1, 2, 3))           # 6
print(toplam(10, 20, 30, 40))    # 100
```

```python
# *args tuple olarak gelir
def goster(*args):
    print(f"Tip: {type(args)}")
    print(f"Değerler: {args}")

goster("a", "b", "c")
```

## **8.5 **kwargs - Sınırsız İsimli Argüman**

```python
# **kwargs ile sınırsız isimli argüman
def kisi_bilgisi(**kwargs):
    for anahtar, deger in kwargs.items():
        print(f"{anahtar}: {deger}")

kisi_bilgisi(ad="Züber", soyad="Doğan", yas=46)
```

```python
# args ve kwargs birlikte
def genel_fonksiyon(*args, **kwargs):
    print("Pozisyonel args:", args)
    print("İsimli kwargs:", kwargs)

genel_fonksiyon(1, 2, 3, isim="Züber", sehir="İstanbul")
```

## **8.6 Pratik Örnek: Maliyet Hesaplama**

```python
def maliyet_hesapla():
    """Ulaşım maliyeti hesaplayan fonksiyon"""
    try:
        gun = int(input("Kaç gün çalışıyorsunuz? "))
        gidis = int(input("Gidiş ücreti: "))
        gelis = int(input("Geliş ücreti: "))
        
        maliyet = gun * (gidis + gelis)
        
        if maliyet <= 100:
            return maliyet, "Düşük maliyet"
        elif maliyet <= 200:
            return maliyet, "Orta maliyet"
        elif maliyet <= 300:
            return maliyet, "Yüksek maliyet"
        else:
            return maliyet, "Acilen çözüm bul!"
            
    except ValueError:
        return 0, "Hata: Sadece sayı girin!"

tutar, durum = maliyet_hesapla()
print(f"\nToplam: {tutar} TL - {durum}")
```
