Parametrelere **varsayÄ±lan deÄŸer** atanabilir. Bu durumda argÃ¼man verilmezse varsayÄ±lan kullanÄ±lÄ±r.

````python
# VarsayÄ±lan deÄŸerli parametre
# ArgÃ¼man verilmezse varsayÄ±lan deÄŸer kullanÄ±lÄ±r

def selamla(isim, mesaj="HoÅŸ geldiniz"):
    """Ä°sim ve mesaj ile selamlar."""
    print(f"{isim}, {mesaj}!")

# Sadece zorunlu parametre
selamla("ZÃ¼ber")  # mesaj varsayÄ±lan = "HoÅŸ geldiniz"

# Her iki parametre
selamla("Uygar", "GÃ¼naydÄ±n")  # mesaj = "GÃ¼naydÄ±n"

# Ä°simli argÃ¼man ile
selamla(mesaj="Ä°yi akÅŸamlar", isim="Ä°dil")
````

````python
# Birden fazla varsayÄ±lan deÄŸerli parametre
# Sondaki parametrelere varsayÄ±lan deÄŸer atanabilir

def us_hesapla(taban, us=2):
    """Ãœs hesaplama. VarsayÄ±lan us=2 (kare)."""
    sonuc = taban ** us
    print(f"{taban}^{us} = {sonuc}")
    return sonuc

# VarsayÄ±lan Ã¼s (2) ile
us_hesapla(5)      # 5^2 = 25
us_hesapla(3)      # 3^2 = 9

# FarklÄ± Ã¼s ile
us_hesapla(2, 10)  # 2^10 = 1024
us_hesapla(5, 3)   # 5^3 = 125
````

# **8.5 return Ä°fadesi**

`return` ifadesi ile fonksiyondan **deÄŸer dÃ¶ndÃ¼rÃ¼lÃ¼r**.

- Fonksiyon Ã§aÄŸrÄ±ldÄ±ÄŸÄ±nda dÃ¶nen deÄŸer kullanÄ±labilir
- `return` olmayan fonksiyon `None` dÃ¶ndÃ¼rÃ¼r
- `return` ifadesi fonksiyonu sonlandÄ±rÄ±r

````python
# print() vs return farkÄ±
# print() ekrana yazar, return deÄŸer dÃ¶ndÃ¼rÃ¼r

def topla_print(a, b):
    """ToplamÄ± ekrana yazdÄ±rÄ±r (deÄŸer dÃ¶ndÃ¼rmez)."""
    print(a + b)  # Ekrana yazar, ama dÃ¶ndÃ¼rmez

def topla_return(a, b):
    """ToplamÄ± dÃ¶ndÃ¼rÃ¼r."""
    return a + b  # DeÄŸeri dÃ¶ndÃ¼rÃ¼r

# print() ile - sonuÃ§ kullanÄ±lamaz
sonuc1 = topla_print(3, 5)  # Ekrana 8 yazar
print(f"sonuc1 = {sonuc1}")  # None! (deÄŸer dÃ¶nmedi)

print()

# return ile - sonuÃ§ kullanÄ±labilir
sonuc2 = topla_return(3, 5)  # 8 dÃ¶ner
print(f"sonuc2 = {sonuc2}")  # 8
print(f"sonuc2 * 2 = {sonuc2 * 2}")  # 16
````

````python
# Tek deÄŸer dÃ¶ndÃ¼rme
# Matematiksel hesaplama Ã¶rneÄŸi

def alan_hesapla(uzunluk, genislik):
    """DikdÃ¶rtgen alanÄ± hesaplar."""
    alan = uzunluk * genislik
    return alan  # Hesaplanan deÄŸeri dÃ¶ndÃ¼r

# DÃ¶nen deÄŸeri kullan
sonuc = alan_hesapla(5, 3)
print(f"Alan: {sonuc} mÂ²")

# DoÄŸrudan baÅŸka iÅŸlemde kullan
toplam_alan = alan_hesapla(10, 20) + alan_hesapla(5, 5)
print(f"Toplam alan: {toplam_alan} mÂ²")
````

````python
# Ã‡oklu deÄŸer dÃ¶ndÃ¼rme
# Birden fazla deÄŸer tuple olarak dÃ¶ner

def dikdortgen_hesapla(uzunluk, genislik):
    """Alan ve Ã§evre hesaplar."""
    alan = uzunluk * genislik
    cevre = 2 * (uzunluk + genislik)
    return alan, cevre  # Ä°ki deÄŸer dÃ¶ndÃ¼r (tuple)

# Tuple olarak al
sonuc = dikdortgen_hesapla(5, 3)
print(f"SonuÃ§ (tuple): {sonuc}")
print(f"Tip: {type(sonuc)}")

# Unpacking ile ayrÄ± deÄŸiÅŸkenlere al
alan, cevre = dikdortgen_hesapla(10, 4)
print(f"\nAlan: {alan} mÂ²")
print(f"Ã‡evre: {cevre} m")
````

````python
# return fonksiyonu sonlandÄ±rÄ±r
# return'den sonraki kod Ã§alÄ±ÅŸmaz!

def pozitif_mi(sayi):
    """SayÄ±nÄ±n pozitif olup olmadÄ±ÄŸÄ±nÄ± dÃ¶ndÃ¼rÃ¼r."""
    if sayi > 0:
        return True   # Buradan Ã§Ä±kÄ±lÄ±r
    elif sayi < 0:
        return False  # Buradan Ã§Ä±kÄ±lÄ±r
    else:
        return None   # 0 iÃ§in
    
    print("Bu satÄ±r asla Ã§alÄ±ÅŸmaz!")  # UlaÅŸÄ±lamaz kod

# Test
print(pozitif_mi(5))   # True
print(pozitif_mi(-3))  # False
print(pozitif_mi(0))   # None
````

# **8.6 *args - SÄ±nÄ±rsÄ±z Pozisyonel ArgÃ¼man**

`*args` ile fonksiyona **sÄ±nÄ±rsÄ±z sayÄ±da** pozisyonel argÃ¼man geÃ§ilebilir.

- ArgÃ¼manlar **tuple** olarak gelir
- `args` ismi gelenekseldir, farklÄ± isim kullanÄ±labilir
- YÄ±ldÄ±z (`*`) zorunludur

````python
# *args ile sÄ±nÄ±rsÄ±z argÃ¼man
# Gelen argÃ¼manlar tuple olarak args'a atanÄ±r

def toplam(*args):
    """SÄ±nÄ±rsÄ±z sayÄ±da argÃ¼manÄ± toplar."""
    print(f"args tipi: {type(args)}")
    print(f"args deÄŸeri: {args}")
    
    sonuc = 0
    for sayi in args:  # Tuple Ã¼zerinde dÃ¶ngÃ¼
        sonuc += sayi
    return sonuc

# FarklÄ± sayÄ±da argÃ¼manla Ã§aÄŸÄ±r
print(f"Toplam: {toplam(1, 2, 3)}")
print(f"Toplam: {toplam(10, 20, 30, 40, 50)}")
print(f"Toplam: {toplam(5)}")
````

````python
# Normal parametre + *args birlikte kullanÄ±mÄ±
# *args en sonda olmalÄ±

def ogrenci_bilgisi(sinif, *ogrenciler):
    """SÄ±nÄ±f ve Ã¶ÄŸrenci listesini gÃ¶sterir."""
    print(f"\nSÄ±nÄ±f: {sinif}")
    print("Ã–ÄŸrenciler:")
    for i, ogrenci in enumerate(ogrenciler, 1):
        print(f"  {i}. {ogrenci}")

# Ã‡aÄŸÄ±r
ogrenci_bilgisi("10-A", "Ali", "Veli", "AyÅŸe")
ogrenci_bilgisi("11-B", "Mehmet", "Fatma")
````

# **8.7 **kwargs - SÄ±nÄ±rsÄ±z Ä°simli ArgÃ¼man**

`**kwargs` ile fonksiyona **sÄ±nÄ±rsÄ±z sayÄ±da** isimli argÃ¼man geÃ§ilebilir.

- ArgÃ¼manlar **dictionary** olarak gelir
- `kwargs` ismi gelenekseldir (keyword arguments)
- Ä°ki yÄ±ldÄ±z (`**`) zorunludur

````python
# **kwargs ile sÄ±nÄ±rsÄ±z isimli argÃ¼man
# Gelen argÃ¼manlar dict olarak kwargs'a atanÄ±r

def kisi_bilgisi(**kwargs):
    """Ä°simli argÃ¼manlarÄ± gÃ¶sterir."""
    print(f"kwargs tipi: {type(kwargs)}")
    print(f"kwargs deÄŸeri: {kwargs}")
    print()
    
    for anahtar, deger in kwargs.items():
        print(f"{anahtar}: {deger}")

# Ä°simli argÃ¼manlarla Ã§aÄŸÄ±r
kisi_bilgisi(ad="ZÃ¼ber", soyad="DoÄŸan", yas=46)
print()
kisi_bilgisi(isim="Python", surum=3.12, tip="Programlama Dili")
````

````python
# *args ve **kwargs birlikte kullanÄ±mÄ±
# SÄ±ra: normal â†’ *args â†’ **kwargs

def genel_fonksiyon(zorunlu, *args, **kwargs):
    """TÃ¼m parametre tiplerini gÃ¶sterir."""
    print(f"Zorunlu: {zorunlu}")
    print(f"*args: {args}")
    print(f"**kwargs: {kwargs}")

# Ã‡aÄŸÄ±r
genel_fonksiyon(
    "Ä°lk parametre",     # zorunlu
    1, 2, 3,             # *args
    isim="ZÃ¼ber",        # **kwargs
    sehir="Ä°stanbul"     # **kwargs
)
````

# **8.8 Docstring (Belgeleme)**

Fonksiyonun ne yaptÄ±ÄŸÄ±nÄ± aÃ§Ä±klayan **Ã¼Ã§ tÄ±rnaklÄ± metin** dokÃ¼man dizisidir.

- Ä°lk satÄ±r kÄ±sa aÃ§Ä±klama
- `help()` ile gÃ¶rÃ¼ntÃ¼lenir
- `__doc__` niteliÄŸi ile eriÅŸilir

````python
# Docstring Ã¶rneÄŸi
# Triple quote iÃ§inde fonksiyon aÃ§Ä±klamasÄ±

def faktoriyel(n):
    """
    Verilen sayÄ±nÄ±n faktÃ¶riyelini hesaplar.
    
    Parametreler:
        n (int): FaktÃ¶riyeli hesaplanacak pozitif tam sayÄ±
    
    DÃ¶nÃ¼ÅŸ:
        int: n! deÄŸeri
    
    Ã–rnek:
        >>> faktoriyel(5)
        120
    """
    if n <= 1:
        return 1
    else:
        return n * faktoriyel(n - 1)

# Fonksiyonu kullan
print(f"5! = {faktoriyel(5)}")

# Docstring'e eriÅŸ
print("\n=== DokÃ¼mantasyon ===")
print(faktoriyel.__doc__)

# help() ile
# help(faktoriyel)
````

# **8.9 Pratik Ã–rnekler**

````python
# Pratik Ã–rnek 1: Hesap Makinesi
# Temel matematiksel iÅŸlemler

def topla(a, b):
    """Ä°ki sayÄ±yÄ± toplar."""
    return a + b

def cikar(a, b):
    """Ä°ki sayÄ±nÄ±n farkÄ±nÄ± alÄ±r."""
    return a - b

def carp(a, b):
    """Ä°ki sayÄ±yÄ± Ã§arpar."""
    return a * b

def bol(a, b):
    """Ä°ki sayÄ±yÄ± bÃ¶ler."""
    if b == 0:
        return "Hata: SÄ±fÄ±ra bÃ¶lÃ¼nemez!"
    return a / b

# Ana hesap makinesi fonksiyonu
def hesap_makinesi(islem, a, b):
    """Ä°ÅŸlem tipine gÃ¶re hesaplama yapar."""
    if islem == "+":
        return topla(a, b)
    elif islem == "-":
        return cikar(a, b)
    elif islem == "*":
        return carp(a, b)
    elif islem == "/":
        return bol(a, b)
    else:
        return "GeÃ§ersiz iÅŸlem!"

# Test
print(f"10 + 5 = {hesap_makinesi('+', 10, 5)}")
print(f"10 - 5 = {hesap_makinesi('-', 10, 5)}")
print(f"10 * 5 = {hesap_makinesi('*', 10, 5)}")
print(f"10 / 5 = {hesap_makinesi('/', 10, 5)}")
print(f"10 / 0 = {hesap_makinesi('/', 10, 0)}")
````

````python
# Pratik Ã–rnek 2: Not OrtalamasÄ± Hesaplama
# *args ile sÄ±nÄ±rsÄ±z not kabul eder

def not_ortalamasi(*notlar):
    """
    Verilen notlarÄ±n ortalamasÄ±nÄ± hesaplar.
    
    DÃ¶nÃ¼ÅŸ: (ortalama, harf_notu, durum)
    """
    if not notlar:
        return 0, "F", "GeÃ§ersiz"
    
    ortalama = sum(notlar) / len(notlar)
    
    # Harf notu belirle
    if ortalama >= 90:
        harf = "AA"
    elif ortalama >= 80:
        harf = "BA"
    elif ortalama >= 70:
        harf = "BB"
    elif ortalama >= 60:
        harf = "CB"
    elif ortalama >= 50:
        harf = "CC"
    else:
        harf = "FF"
    
    # Durum belirle
    durum = "GeÃ§ti" if ortalama >= 50 else "KaldÄ±"
    
    return ortalama, harf, durum

# Test
ort, harf, durum = not_ortalamasi(85, 90, 78, 92)
print(f"Ortalama: {ort:.2f}")
print(f"Harf Notu: {harf}")
print(f"Durum: {durum}")

print()

ort, harf, durum = not_ortalamasi(40, 35, 45)
print(f"Ortalama: {ort:.2f}")
print(f"Harf Notu: {harf}")
print(f"Durum: {durum}")
````

````python
# Pratik Ã–rnek 3: Asal SayÄ± KontrolÃ¼
# Fonksiyonlar arasÄ± iÅŸbirliÄŸi

def asal_mi(sayi):
    """SayÄ±nÄ±n asal olup olmadÄ±ÄŸÄ±nÄ± kontrol eder."""
    if sayi < 2:
        return False
    if sayi == 2:
        return True
    if sayi % 2 == 0:
        return False
    
    for i in range(3, int(sayi ** 0.5) + 1, 2):
        if sayi % i == 0:
            return False
    return True

def asal_sayilar_bul(baslangic, bitis):
    """AralÄ±ktaki asal sayÄ±larÄ± bulur."""
    asallar = []
    for sayi in range(baslangic, bitis + 1):
        if asal_mi(sayi):  # BaÅŸka fonksiyonu Ã§aÄŸÄ±r
            asallar.append(sayi)
    return asallar

# Test
print(f"7 asal mÄ±? {asal_mi(7)}")
print(f"12 asal mÄ±? {asal_mi(12)}")
print(f"\n1-50 arasÄ± asallar: {asal_sayilar_bul(1, 50)}")
````

````python
# ============================================================================
# ğŸŸ¨ CEVAP ANAHTARI â€“ Fonksiyonlar
# ============================================================================
# YukarÄ±daki gÃ¶revlerin tamamlanmÄ±ÅŸ Ã§Ã¶zÃ¼mleri aÅŸaÄŸÄ±da verilmiÅŸtir.
# ============================================================================




# ============================================================================
# ğŸŸ¨ CEVAP 1: Basit fonksiyon tanÄ±mlama ve Ã§aÄŸÄ±rma
# ============================================================================
def merhaba():
    print("Merhaba Python!")
    print("Fonksiyon Ã§alÄ±ÅŸtÄ±.")
merhaba()

print(f"\n{'-'*82}\n")




# ============================================================================
# ğŸŸ¨ CEVAP 2: Fonksiyonu birden fazla Ã§aÄŸÄ±rma
# ============================================================================
def yatay_cizgi():
    print("=" * 40)
yatay_cizgi()
print("BAÅLIK")
yatay_cizgi()
print("Ä°Ã§erik")
yatay_cizgi()

print(f"\n{'-'*82}\n")




# ============================================================================
# ğŸŸ¨ CEVAP 3: Tek parametreli fonksiyon
# ============================================================================
def hosgeldin(isim):
    print(f"HoÅŸ geldin, {isim}!")
hosgeldin("Ali")
hosgeldin("AyÅŸe")
hosgeldin("Mehmet")

print(f"\n{'-'*82}\n")




# ============================================================================
# ğŸŸ¨ CEVAP 4: Ã‡oklu parametreli fonksiyon
# ============================================================================
def urun_bilgisi(urun, fiyat, adet):
    print(f"ÃœrÃ¼n: {urun}")
    print(f"Fiyat: {fiyat} TL")
    print(f"Adet: {adet}")
    print("-" * 20)
urun_bilgisi("Kalem", 15, 100)
urun_bilgisi("Defter", 30, 50)

print(f"\n{'-'*82}\n")




# ============================================================================
# ğŸŸ¨ CEVAP 5: Ä°simli argÃ¼manlar (keyword arguments)
# ============================================================================
def ogrenci_goster(ad, sinif, okul):
    print(f"{ad}, {sinif} sÄ±nÄ±fÄ±, {okul}")
ogrenci_goster("Ali", "10-A", "AtatÃ¼rk Lisesi")
ogrenci_goster(okul="MYO", ad="Veli", sinif="1-B")

print(f"\n{'-'*82}\n")




# ============================================================================
# ğŸŸ¨ CEVAP 6: VarsayÄ±lan parametre deÄŸerleri
# ============================================================================
def selam(isim, dil="TÃ¼rkÃ§e"):
    if dil == "TÃ¼rkÃ§e":
        print(f"Merhaba {isim}!")
    else:
        print(f"Hello {isim}!")
selam("Ali")                    # Merhaba Ali!
selam("John", "Ä°ngilizce")     # Hello John!

print(f"\n{'-'*82}\n")




# ============================================================================
# ğŸŸ¨ CEVAP 7: VarsayÄ±lan deÄŸerli Ã¼s hesaplama
# ============================================================================
def kuvvet(taban, us=3):
    return taban ** us
print(kuvvet(2))       # 8
print(kuvvet(5))       # 125
print(kuvvet(2, 10))   # 1024

print(f"\n{'-'*82}\n")




# ============================================================================
# ğŸŸ¨ CEVAP 8: print() vs return farkÄ±
# ============================================================================
def carp_print(a, b):
    print(a * b)
def carp_return(a, b):
    return a * b
sonuc1 = carp_print(3, 4)   # 12 yazdÄ±rÄ±r
sonuc2 = carp_return(3, 4)  # 12 dÃ¶ndÃ¼rÃ¼r
print(f"print sonucu: {sonuc1}")   # None
print(f"return sonucu: {sonuc2}")  # 12

print(f"\n{'-'*82}\n")




# ============================================================================
# ğŸŸ¨ CEVAP 9: Tek deÄŸer dÃ¶ndÃ¼rme
# ============================================================================
def ucgen_alan(taban, yukseklik):
    return (taban * yukseklik) / 2
sonuc = ucgen_alan(10, 5)
print(f"ÃœÃ§gen alanÄ±: {sonuc}")  # 25.0

print(f"\n{'-'*82}\n")




# ============================================================================
# ğŸŸ¨ CEVAP 10: Ã‡oklu deÄŸer dÃ¶ndÃ¼rme
# ============================================================================
def daire_hesapla(yaricap):
    alan = 3.14 * yaricap ** 2
    cevre = 2 * 3.14 * yaricap
    return alan, cevre
a, c = daire_hesapla(7)
print(f"Alan: {a:.2f}")    # 153.86
print(f"Ã‡evre: {c:.2f}")  # 43.96

print(f"\n{'-'*82}\n")




# ============================================================================
# ğŸŸ¨ CEVAP 11: *args â€“ SÄ±nÄ±rsÄ±z pozisyonel argÃ¼man
# ============================================================================
def ortalama(*args):
    return sum(args) / len(args)
print(ortalama(10, 20, 30))          # 20.0
print(ortalama(5, 15, 25, 35, 45))   # 25.0

print(f"\n{'-'*82}\n")




# ============================================================================
# ğŸŸ¨ CEVAP 12: **kwargs â€“ SÄ±nÄ±rsÄ±z isimli argÃ¼man
# ============================================================================
def profil(**kwargs):
    for anahtar, deger in kwargs.items():
        print(f"{anahtar}: {deger}")
profil(ad="ZÃ¼ber", yas=46, sehir="Ä°stanbul")

print(f"\n{'-'*82}\n")




# ============================================================================
# ğŸŸ¨ CEVAP 13: Docstring kullanÄ±mÄ±
# ============================================================================
def faktoriyel(n):
    """
    Verilen sayÄ±nÄ±n faktÃ¶riyelini hesaplar.
    Parametre: n (int)
    DÃ¶nÃ¼ÅŸ: int
    """
    if n <= 1:
        return 1
    return n * faktoriyel(n - 1)
print(f"5! = {faktoriyel(5)}")  # 120
print(faktoriyel.__doc__)

print(f"\n{'-'*82}\n")




# ============================================================================
# ğŸŸ¨ CEVAP 14: Pratik Ã–rnek â€“ Hesap makinesi
# ============================================================================
def topla(a, b): return a + b
def cikar(a, b): return a - b
def carp(a, b): return a * b
def bol(a, b):
    if b == 0: return "SÄ±fÄ±ra bÃ¶lÃ¼nemez!"
    return a / b
def hesapla(islem, a, b):
    if islem == "+": return topla(a, b)
    elif islem == "-": return cikar(a, b)
    elif islem == "*": return carp(a, b)
    elif islem == "/": return bol(a, b)
print(hesapla("+", 15, 5))   # 20
print(hesapla("*", 6, 7))    # 42
print(hesapla("/", 10, 0))   # SÄ±fÄ±ra bÃ¶lÃ¼nemez!

print(f"\n{'-'*82}\n")




# ============================================================================
# ğŸŸ¨ CEVAP 15: Pratik Ã–rnek â€“ Asal sayÄ± kontrolÃ¼
# ============================================================================
def asal_mi(sayi):
    if sayi < 2: return False
    if sayi == 2: return True
    if sayi % 2 == 0: return False
    for i in range(3, int(sayi**0.5)+1, 2):
        if sayi % i == 0: return False
    return True
asallar = []
for s in range(1, 31):
    if asal_mi(s):
        asallar.append(s)
print(f"1-30 arasÄ± asallar: {asallar}")
````
