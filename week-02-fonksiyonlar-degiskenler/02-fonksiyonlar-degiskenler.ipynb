# **2.1 DeÄŸiÅŸkenler**

Python'da bir program iÃ§inde deÄŸerlere verilen isimlere **deÄŸiÅŸken** denir. DeÄŸiÅŸkenler, verileri geÃ§ici olarak bellekte saklamamÄ±zÄ± ve bu verilere istediÄŸimiz zaman eriÅŸmemizi saÄŸlar.

DeÄŸiÅŸken kullanmanÄ±n temel avantajlarÄ±:

### DeÄŸiÅŸken TanÄ±mlama

Python'da deÄŸiÅŸken tanÄ±mlamak iÃ§in `=` (atama operatÃ¶rÃ¼) kullanÄ±lÄ±r. Atama operatÃ¶rÃ¼, saÄŸ taraftaki deÄŸeri sol taraftaki deÄŸiÅŸkene atar.

````python
# DeÄŸiÅŸken tanÄ±mlama - Atama operatÃ¶rÃ¼ (=) kullanÄ±mÄ±
# SaÄŸ taraftaki deÄŸer, sol taraftaki deÄŸiÅŸkene atanÄ±r
n = 5  # n adlÄ± deÄŸiÅŸkene 5 deÄŸeri atanÄ±yor
print(n)  # DeÄŸiÅŸkenin deÄŸerini ekrana yazdÄ±rÄ±r: 5

# DeÄŸiÅŸken kullanarak iÅŸlem yapma
# DeÄŸiÅŸken adÄ± kullanÄ±ldÄ±ÄŸÄ±nda, Python deÄŸeri otomatik yerine koyar
print(n * 10)  # 5 * 10 = 50
print(n / 2)   # 5 / 2 = 2.5
````

````python
# Birden fazla deÄŸiÅŸken tanÄ±mlama
# Her deÄŸiÅŸken ayrÄ± bir satÄ±rda tanÄ±mlanabilir
isim = "ZÃ¼ber"      # String (karakter dizisi) deÄŸer
soyisim = "DoÄŸan"   # String deÄŸer
yas = 46            # Integer (tam sayÄ±) deÄŸer

# DeÄŸiÅŸkenleri print() ile yazdÄ±rma
# Birden fazla deÄŸiÅŸken virgÃ¼lle ayrÄ±larak tek satÄ±rda yazdÄ±rÄ±labilir
print(isim, soyisim, yas)  # ZÃ¼ber DoÄŸan 46
````

### DeÄŸiÅŸken Ä°simlendirme KurallarÄ±

| Kural | GeÃ§erli | GeÃ§ersiz |

**Tavsiye Edilen Kurallar:**

````python
# Ã‡oklu deÄŸer atama - AynÄ± deÄŸeri birden fazla deÄŸiÅŸkene atama
# Birden fazla deÄŸiÅŸken tek satÄ±rda aynÄ± deÄŸere eÅŸitlenebilir
# Bu yÃ¶ntem, aynÄ± deÄŸeri paylaÅŸan deÄŸiÅŸkenler iÃ§in pratiktir
ocak = mart = mayis = temmuz = agustos = ekim = aralik = 31  # 31 Ã§ekilen aylar
nisan = haziran = eylul = kasim = 30  # 30 Ã§ekilen aylar
subat = 28  # Åubat (artÄ±k yÄ±l hariÃ§)

print("Ocak:", ocak, "gÃ¼n")    # Ocak: 31 gÃ¼n
print("Nisan:", nisan, "gÃ¼n")  # Nisan: 30 gÃ¼n
print("Åubat:", subat, "gÃ¼n")  # Åubat: 28 gÃ¼n
````

````python
# DeÄŸiÅŸken deÄŸerlerini takas etme (swap)
# Python'da iki deÄŸiÅŸkenin deÄŸerini tek satÄ±rda takas edebilirsiniz
# Bu Ã¶zellik Python'a Ã¶zgÃ¼dÃ¼r ve geÃ§ici deÄŸiÅŸken gerektirmez
osman = "AraÅŸtÄ±rma GeliÅŸtirme MÃ¼dÃ¼rÃ¼"
mehmet = "Proje Sorumlusu"

print("Takas Ã¶ncesi:")
print("Osman:", osman)   # AraÅŸtÄ±rma GeliÅŸtirme MÃ¼dÃ¼rÃ¼
print("Mehmet:", mehmet) # Proje Sorumlusu

# Tek satÄ±rda deÄŸer takasÄ± - Python'Ä±n gÃ¼Ã§lÃ¼ Ã¶zelliklerinden biri
osman, mehmet = mehmet, osman  # DeÄŸerler yer deÄŸiÅŸtirdi

print("\nTakas sonrasÄ±:")
print("Osman:", osman)   # Proje Sorumlusu
print("Mehmet:", mehmet) # AraÅŸtÄ±rma GeliÅŸtirme MÃ¼dÃ¼rÃ¼
````

### input() KullanÄ±mÄ±


````python
# input() fonksiyonu ile kullanÄ±cÄ±dan veri alma
# Parantez iÃ§indeki metin, kullanÄ±cÄ±ya soru olarak gÃ¶sterilir
# KullanÄ±cÄ±nÄ±n girdiÄŸi deÄŸer, deÄŸiÅŸkene atanÄ±r
isim = input("AdÄ±nÄ±zÄ± girin: ")  # KullanÄ±cÄ±dan isim ister

# KullanÄ±cÄ±nÄ±n girdiÄŸi deÄŸeri kullanma
print("Merhaba", isim)  # Merhaba [kullanÄ±cÄ±nÄ±n girdiÄŸi isim]
````

# **2.3 Tip DÃ¶nÃ¼ÅŸÃ¼mleri**

Python'da **o anda elimizde bulunan verinin hangi tipte olduÄŸunu bilmek son derece Ã¶nemlidir**. Ã‡Ã¼nkÃ¼ bir verinin ait olduÄŸu tip, o veriyle neler yapÄ±p neler yapamayacaÄŸÄ±nÄ±zÄ± belirler.

Aritmetik iÅŸlemler yalnÄ±zca sayÄ±larla yapÄ±lÄ±r. Karakter dizileri ile aritmetik iÅŸlem yapÄ±lamaz. Bu nedenle `input()` fonksiyonundan gelen verilerin tip dÃ¶nÃ¼ÅŸÃ¼mÃ¼ yapÄ±lmalÄ±dÄ±r.

---

### Tip DÃ¶nÃ¼ÅŸÃ¼m FonksiyonlarÄ±

| Fonksiyon | Ä°ÅŸlev | Ã–rnek |
|-----------|-------|-------|
| `int()` | Tam sayÄ±ya Ã§evirir | `int("23")` â†’ `23` |
| `float()` | OndalÄ±klÄ± sayÄ±ya Ã§evirir | `float("23.5")` â†’ `23.5` |
| `str()` | Karakter dizisine Ã§evirir | `str(2024)` â†’ `"2024"` |

````python
# int() - String'i tam sayÄ±ya Ã§evirme
# KullanÄ±m: int(karakter_dizisi)
# Dikkat: Karakter dizisi geÃ§erli bir tam sayÄ± iÃ§ermelidir
sayi_str = "23"  # Bu bir karakter dizisi (string)
sayi_int = int(sayi_str)  # Tam sayÄ±ya (integer) Ã§evrildi

print("Orijinal deÄŸer:", sayi_str, "- Tip:", type(sayi_str))  # str
print("DÃ¶nÃ¼ÅŸtÃ¼rÃ¼lmÃ¼ÅŸ deÄŸer:", sayi_int, "- Tip:", type(sayi_int))  # int
````

````python
# float() - String'i ondalÄ±klÄ± sayÄ±ya Ã§evirme
# KullanÄ±m: float(karakter_dizisi)
# OndalÄ±k ayÄ±rÄ±cÄ± olarak nokta (.) kullanÄ±lmalÄ±dÄ±r
sayi_str = "23.5"  # OndalÄ±klÄ± sayÄ± iÃ§eren string
sayi_float = float(sayi_str)  # Float'a Ã§evrildi

print("Orijinal deÄŸer:", sayi_str, "- Tip:", type(sayi_str))  # str
print("DÃ¶nÃ¼ÅŸtÃ¼rÃ¼lmÃ¼ÅŸ deÄŸer:", sayi_float, "- Tip:", type(sayi_float))  # float
````

````python
# str() - SayÄ±yÄ± karakter dizisine Ã§evirme
# KullanÄ±m: str(sayi)
# SayÄ±larÄ± metinlerle birleÅŸtirmek iÃ§in gereklidir
yil = 2024  # Tam sayÄ±
yil_str = str(yil)  # String'e Ã§evrildi

print("Orijinal deÄŸer:", yil, "- Tip:", type(yil))  # int
print("DÃ¶nÃ¼ÅŸtÃ¼rÃ¼lmÃ¼ÅŸ deÄŸer:", yil_str, "- Tip:", type(yil_str))  # str

# str() kullanarak metin birleÅŸtirme
# SayÄ±larÄ± + ile metne eklemek iÃ§in str() gerekir
mesaj = "YÄ±l: " + yil_str  # + ile birleÅŸtirme
print(mesaj)  # YÄ±l: 2024
````

````python
# Tip dÃ¶nÃ¼ÅŸÃ¼mÃ¼ olmadan hata Ã¶rneÄŸi
# Karakter dizileri ile aritmetik iÅŸlem yapÄ±lamaz!
a = "10"
b = "20"

# + iÅŸareti string'leri birleÅŸtirir, toplamaz!
print("String toplama:", a + b)  # 1020 (birleÅŸtirme)

# DoÄŸru yol: Ã–nce tip dÃ¶nÃ¼ÅŸÃ¼mÃ¼, sonra iÅŸlem
print("SayÄ± toplama:", int(a) + int(b))  # 30 (gerÃ§ek toplama)
````

# **2.4 format() ve f-string**

Python'da karakter dizilerini biÃ§imlendirmek iÃ§in Ã§eÅŸitli yÃ¶ntemler bulunur. En yaygÄ±n kullanÄ±lanlar `format()` metodu ve **f-string** (formatted string literals) yÃ¶ntemidir.

---

### format() Metodu

`format()` metodu, karakter dizisi iÃ§indeki `{}` yer tutucularÄ±nÄ± belirtilen deÄŸerlerle deÄŸiÅŸtirir.

### f-string (Python 3.6+)

f-string, modern Python'da tercih edilen biÃ§imlendirme yÃ¶ntemidir. Karakter dizisinin baÅŸÄ±na `f` harfi eklenir ve deÄŸiÅŸkenler doÄŸrudan `{}` iÃ§inde yazÄ±lÄ±r.

````python
# format() metodu kullanÄ±mÄ±
# {} yer tutucularÄ±, format() iÃ§indeki deÄŸerlerle sÄ±rayla deÄŸiÅŸtirilir
# Ä°lk {} ilk parametreyle, ikinci {} ikinci parametreyle eÅŸleÅŸir
print("{} ve {} BeÅŸiktaÅŸlÄ±dÄ±r".format("Uygar", "ZÃ¼ber"))  # Uygar ve ZÃ¼ber BeÅŸiktaÅŸlÄ±dÄ±r
````

````python
# format() ile indeks kullanÄ±mÄ±
# {0}, {1} gibi indekslerle parametre sÄ±rasÄ± belirlenebilir
# AynÄ± deÄŸer birden fazla yerde kullanÄ±labilir
print("{0} {1} yaÅŸÄ±nda. {0} Ankara'da yaÅŸÄ±yor.".format("Ali", 25))
# Ali 25 yaÅŸÄ±nda. Ali Ankara'da yaÅŸÄ±yor.
````

````python
# f-string kullanÄ±mÄ± (Python 3.6+ iÃ§in Ã¶nerilen yÃ¶ntem)
# String'in baÅŸÄ±na 'f' harfi eklenir
# DeÄŸiÅŸkenler doÄŸrudan {} iÃ§inde yazÄ±lÄ±r
isim1 = "Uygar"
isim2 = "ZÃ¼ber"

# f-string ile biÃ§imlendirme - en okunabilir yÃ¶ntem
print(f"{isim1} ve {isim2} BeÅŸiktaÅŸlÄ±dÄ±r")  # Uygar ve ZÃ¼ber BeÅŸiktaÅŸlÄ±dÄ±r
````

````python
# f-string iÃ§inde ifade kullanÄ±mÄ±
# {} iÃ§inde matematiksel iÅŸlemler ve fonksiyonlar kullanÄ±labilir
yas = 46
ad = "ZÃ¼ber"

# Matematiksel iÅŸlem f-string iÃ§inde
print(f"{ad}, 10 yÄ±l sonra {yas + 10} yaÅŸÄ±nda olacak.")  # ZÃ¼ber, 10 yÄ±l sonra 56 yaÅŸÄ±nda olacak.

# Fonksiyon Ã§aÄŸrÄ±sÄ± f-string iÃ§inde
metin = "python"
print(f"{metin} kelimesi {len(metin)} harften oluÅŸur.")  # python kelimesi 6 harften oluÅŸur.
````

# **2.5 round() ve bool() FonksiyonlarÄ±**

Python'da sÄ±k kullanÄ±lan iki gÃ¶mÃ¼lÃ¼ fonksiyon daha vardÄ±r: `round()` (yuvarlama) ve `bool()` (Boolean dÃ¶nÃ¼ÅŸÃ¼mÃ¼).

---

### round() Fonksiyonu

OndalÄ±klÄ± sayÄ±larÄ± belirli basamak sayÄ±sÄ±na yuvarlamak iÃ§in kullanÄ±lÄ±r.

### bool() Fonksiyonu

Herhangi bir deÄŸeri Boolean (`True` veya `False`) deÄŸere dÃ¶nÃ¼ÅŸtÃ¼rÃ¼r. Python'da "boÅŸ" veya "sÄ±fÄ±r" deÄŸerler `False`, diÄŸerleri `True` dÃ¶ndÃ¼rÃ¼r.

````python
# round() fonksiyonu - OndalÄ±klÄ± sayÄ±larÄ± yuvarlama
# KullanÄ±m: round(sayi, basamak_sayisi)
# basamak_sayisi verilmezse tam sayÄ±ya yuvarlar
pi = 22 / 7  # Pi sayÄ±sÄ±nÄ±n yaklaÅŸÄ±k deÄŸeri

print("Pi (orijinal):", pi)           # 3.142857142857143
print("Pi (2 basamak):", round(pi, 2))  # 3.14
print("Pi (3 basamak):", round(pi, 3))  # 3.143
print("Pi (tam sayÄ±):", round(pi))      # 3
````

````python
# bool() fonksiyonu - Boolean deÄŸere dÃ¶nÃ¼ÅŸtÃ¼rme
# BoÅŸ ve sÄ±fÄ±r deÄŸerler False, diÄŸerleri True dÃ¶ndÃ¼rÃ¼r

# SayÄ±lar
print("bool(0):", bool(0))      # False - SÄ±fÄ±r boÅŸ kabul edilir
print("bool(1):", bool(1))      # True - SÄ±fÄ±r olmayan sayÄ±lar
print("bool(-5):", bool(-5))    # True - Negatif sayÄ±lar da True

# Karakter dizileri
print("bool(''):", bool(""))    # False - BoÅŸ string
print("bool('a'):", bool("a"))  # True - Dolu string

# Listeler
print("bool([]):", bool([]))    # False - BoÅŸ liste
print("bool([1]):", bool([1]))  # True - Dolu liste
````

# **2.6 Pratik Ã–rnek: Fatura Hesaplama**

Bu bÃ¶lÃ¼mde Ã¶ÄŸrendiklerimizi birleÅŸtirerek pratik bir Ã¶rnek yapacaÄŸÄ±z: DoÄŸalgaz faturasÄ±nÄ± analiz eden bir program.

````python
# Pratik Ã–rnek: DoÄŸalgaz Fatura Analizi
# Bu Ã¶rnekte deÄŸiÅŸkenler, matematiksel iÅŸlemler ve round() fonksiyonu kullanÄ±lÄ±yor

# Ay bilgileri - sabit deÄŸerler
mart_gun = 31  # Mart ayÄ±nÄ±n gÃ¼n sayÄ±sÄ±
nisan_gun = 30  # Nisan ayÄ±nÄ±n gÃ¼n sayÄ±sÄ±

# Mart ayÄ± fatura verileri
mart_fatura_tl = 273.87  # Mart faturasÄ± (TL)
mart_tuketim_m3 = 346    # Mart tÃ¼ketimi (m3)

# Birim fiyat hesaplama
# FormÃ¼l: Fatura tutarÄ± / TÃ¼ketim miktarÄ± = Birim fiyat
birim_fiyat = mart_fatura_tl / mart_tuketim_m3
print(f"Birim fiyat: {round(birim_fiyat, 2)} TL/m3")  # 0.79 TL/m3

# GÃ¼nlÃ¼k ortalama tÃ¼ketim hesaplama
# FormÃ¼l: AylÄ±k tÃ¼ketim / GÃ¼n sayÄ±sÄ± = GÃ¼nlÃ¼k tÃ¼ketim
gunluk_tuketim = mart_tuketim_m3 / mart_gun
print(f"GÃ¼nlÃ¼k ortalama tÃ¼ketim: {round(gunluk_tuketim, 2)} m3")  # 11.16 m3

# Nisan ayÄ± fatura tahmini
# VarsayÄ±m: Nisan'da gÃ¼nlÃ¼k 11 m3 tÃ¼ketim
nisan_gunluk = 11
nisan_fatura = nisan_gun * nisan_gunluk * birim_fiyat
print(f"Nisan tahmini fatura: {round(nisan_fatura, 2)} TL")  # 261.3 TL
````

````python
# Pratik Ã–rnek: Daire AlanÄ± Hesaplama (input kullanarak)
# Bu Ã¶rnekte input(), int/float dÃ¶nÃ¼ÅŸÃ¼mÃ¼ ve matematiksel iÅŸlemler kullanÄ±lÄ±yor

# KullanÄ±cÄ±dan dairenin Ã§apÄ±nÄ± al
cap = input("Dairenin Ã§apÄ±nÄ± girin (cm): ")

# YarÄ±Ã§ap hesapla - input() string dÃ¶ndÃ¼rdÃ¼ÄŸÃ¼ iÃ§in float() ile dÃ¶nÃ¼ÅŸtÃ¼r
# FormÃ¼l: yarÄ±Ã§ap = Ã§ap / 2
yaricap = float(cap) / 2

# Pi sayÄ±sÄ± (sabit)
pi = 3.14159

# Alan hesaplama
# FormÃ¼l: alan = Ï€ Ã— rÂ²
alan = pi * (yaricap ** 2)  # ** kuvvet alma operatÃ¶rÃ¼

# Sonucu yazdÄ±r
print(f"Ã‡apÄ± {cap} cm olan dairenin alanÄ±: {round(alan, 2)} cmÂ²")
````
