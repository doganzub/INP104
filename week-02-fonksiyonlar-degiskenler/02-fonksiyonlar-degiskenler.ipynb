# **2. Gömülü Fonksiyonlar ve Değişkenler**

Bu derste Python'un gömülü fonksiyonlarını ve değişken tanımlama kurallarını öğreneceğiz.

## **2.1 Tip Dönüşüm Fonksiyonları**

```python
# int() - Tam sayıya çevirme
sayi_str = "23"
sayi_int = int(sayi_str)
print(sayi_int, type(sayi_int))
```

```python
# float() - Ondalıklı sayıya çevirme
sayi = float("23.5")
print(sayi, type(sayi))
```

```python
# str() - Karakter dizisine çevirme
yil = 2024
yil_str = str(yil)
print(yil_str, type(yil_str))
```

## **2.2 Input Fonksiyonu**

Kullanıcıdan veri almak için `input()` fonksiyonu kullanılır. Input her zaman string döndürür!

```python
# Kullanıcıdan veri alma
isim = input("Adınızı girin: ")
print("Merhaba", isim)
```

```python
# Sayısal işlem için tip dönüşümü
gun = input("Kaç gün çalışıyorsunuz? ")
gidis = input("Gidiş ücreti: ")
gelis = input("Geliş ücreti: ")

maliyet = int(gun) * (int(gidis) + int(gelis))
print("Toplam maliyet:", maliyet, "TL")
```

## **2.3 Değişken Tanımlama**

```python
# Temel değişken tanımlama
isim = "Züber"
soyisim = "Doğan"
yas = 46

print(isim, soyisim, yas)
```

```python
# Çoklu değer atama
ocak = mart = mayis = temmuz = agustos = ekim = aralik = 31
nisan = haziran = eylul = kasim = 30
subat = 28

print("Ocak:", ocak, "gün")
print("Nisan:", nisan, "gün")
print("Şubat:", subat, "gün")
```

```python
# Değişken değerlerini değiştirme
osman = "Araştırma Geliştirme Müdürü"
mehmet = "Proje Sorumlusu"

# Değerleri takas etme
osman, mehmet = mehmet, osman

print("Osman:", osman)
print("Mehmet:", mehmet)
```

## **2.4 Format Fonksiyonu**

```python
# format() ile metin biçimlendirme
print("{} ve {} Beşiktaşlıdır".format("Uygar", "Züber"))
```

```python
# f-string kullanımı (modern yöntem)
isim1 = "Uygar"
isim2 = "Züber"
print(f"{isim1} ve {isim2} Beşiktaşlıdır")
```

## **2.5 Round ve Bool Fonksiyonları**

```python
# round() - Yuvarlama
pi = 22/7
print("Pi sayısı:", round(pi, 3))  # 3 ondalık basamak
```

```python
# bool() - Boolean değer
print(bool(0))    # False
print(bool(1))    # True
print(bool(""))   # False (boş string)
print(bool("a"))  # True
```

## **2.6 Pratik Örnek: Fatura Hesaplama**

```python
# Doğalgaz fatura hesaplama örneği
mart_gun = 31
nisan_gun = 30

mart_fatura_tl = 273.87
mart_tuketim_m3 = 346

# Birim fiyat hesaplama
birim_fiyat = mart_fatura_tl / mart_tuketim_m3
print("Birim fiyat:", round(birim_fiyat, 2), "TL/m3")

# Günlük tüketim
gunluk_tuketim = mart_tuketim_m3 / mart_gun
print("Günlük tüketim:", round(gunluk_tuketim, 2), "m3")

# Nisan tahmini (11 m3/gün varsayımıyla)
nisan_gunluk = 11
nisan_fatura = nisan_gun * nisan_gunluk * birim_fiyat
print("Nisan tahmini fatura:", round(nisan_fatura, 2), "TL")
```
