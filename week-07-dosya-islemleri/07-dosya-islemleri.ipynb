# **7.2 Dosya AÃ§ma Kipleri**

`open()` fonksiyonu ile dosya aÃ§Ä±lÄ±r. Ä°kinci parametre **kip (mode)** belirler.

| Kip | AÃ§Ä±klama | Dosya Varsa | Dosya Yoksa |
|-----|----------|-------------|-------------|
| `"r"` | Okuma (varsayÄ±lan) | AÃ§ | Hata |
| `"w"` | Yazma | Ä°Ã§eriÄŸi sil, yeniden oluÅŸtur | OluÅŸtur |
| `"a"` | Ekleme (append) | Sona ekle | OluÅŸtur |
| `"r+"` | Okuma + Yazma | AÃ§ | Hata |
| `"w+"` | Yazma + Okuma | Ä°Ã§eriÄŸi sil | OluÅŸtur |

## SÃ¶zdizimi

```python
dosya = open("dosya_adÄ±", "kip")
```

````python
# Dosya oluÅŸturma ve yazma - "w" kipi
# "w" kipi: Dosya yoksa oluÅŸturur, varsa iÃ§eriÄŸini siler!

# Dosya aÃ§ - yazma kipinde
dosya = open("ornek.txt", "w")  # "w" = write (yazma)

# Dosyaya yaz
dosya.write("Merhaba DÃ¼nya!\n")  # \n = yeni satÄ±r karakteri
dosya.write("Python dosya iÅŸlemleri Ã¶ÄŸreniyorum.\n")
dosya.write("Bu Ã¼Ã§Ã¼ncÃ¼ satÄ±r.\n")

# DosyayÄ± kapat (Ã–NEMLÄ°!)
dosya.close()  # KaynaklarÄ± serbest bÄ±rak, verilerin yazÄ±lmasÄ±nÄ± garantile

print("Dosya oluÅŸturuldu ve yazÄ±ldÄ±!")
````

````python
# Dosya ekleme - "a" kipi
# "a" kipi: Mevcut iÃ§eriÄŸi korur, sona ekler

# Dosya aÃ§ - ekleme kipinde
dosya = open("ornek.txt", "a")  # "a" = append (ekleme)

# DosyanÄ±n sonuna ekle
dosya.write("Bu satÄ±r sona eklendi.\n")
dosya.write("Bu da sona eklendi.\n")

# DosyayÄ± kapat
dosya.close()

print("SatÄ±rlar eklendi!")
````

# **7.3 Dosya Okuma**

Dosya okumak iÃ§in Ã¼Ã§ farklÄ± metot kullanÄ±lÄ±r:

| Metot | AÃ§Ä±klama | DÃ¶nen DeÄŸer |
|-------|----------|-------------|
| `read()` | TÃ¼m dosyayÄ± oku | Karakter dizisi |
| `readline()` | Tek satÄ±r oku | Karakter dizisi |
| `readlines()` | TÃ¼m satÄ±rlarÄ± oku | Liste |

````python
# read() metodu - TÃ¼m dosyayÄ± oku
# DosyanÄ±n tamamÄ±nÄ± tek bir karakter dizisi olarak dÃ¶ndÃ¼rÃ¼r

# Dosya aÃ§ - okuma kipinde
dosya = open("ornek.txt", "r")  # "r" = read (okuma), varsayÄ±lan

# TÃ¼m iÃ§eriÄŸi oku
icerik = dosya.read()  # Karakter dizisi dÃ¶ner

# DosyayÄ± kapat
dosya.close()

# Ä°Ã§eriÄŸi gÃ¶ster
print("=== Dosya Ä°Ã§eriÄŸi ===")
print(icerik)
print(f"\nVeri tipi: {type(icerik)}")
````

````python
# readline() metodu - SatÄ±r satÄ±r okuma
# Her Ã§aÄŸrÄ±da bir sonraki satÄ±rÄ± okur (imleÃ§ ilerler)

# Dosya aÃ§
dosya = open("ornek.txt", "r")

# SatÄ±r satÄ±r oku
satir1 = dosya.readline()  # 1. satÄ±r
satir2 = dosya.readline()  # 2. satÄ±r
satir3 = dosya.readline()  # 3. satÄ±r

# DosyayÄ± kapat
dosya.close()

# strip() ile satÄ±r sonu karakterini (\n) kaldÄ±r
print("1. satÄ±r:", satir1.strip())
print("2. satÄ±r:", satir2.strip())
print("3. satÄ±r:", satir3.strip())
````

````python
# readlines() metodu - TÃ¼m satÄ±rlarÄ± liste olarak al
# Her satÄ±r listenin bir Ã¶ÄŸesi olur

# Dosya aÃ§
dosya = open("ornek.txt", "r")

# TÃ¼m satÄ±rlarÄ± liste olarak al
satirlar = dosya.readlines()  # Liste dÃ¶ner!

# DosyayÄ± kapat
dosya.close()

# Liste iÃ§eriÄŸini gÃ¶ster
print(f"Veri tipi: {type(satirlar)}")
print(f"SatÄ±r sayÄ±sÄ±: {len(satirlar)}")
print("\nListe:")
print(satirlar)

# DÃ¶ngÃ¼ ile satÄ±rlarÄ± gÃ¶ster
print("\nSatÄ±r satÄ±r:")
for i, satir in enumerate(satirlar, 1):
    print(f"{i}. {satir.strip()}")  # strip() ile \n kaldÄ±r
````

````python
# Dosya Ã¼zerinde doÄŸrudan dÃ¶ngÃ¼ kurma
# En bellek verimli yÃ¶ntem - bÃ¼yÃ¼k dosyalar iÃ§in ideal

# Dosya aÃ§
dosya = open("ornek.txt", "r")

# Dosya Ã¼zerinde doÄŸrudan dÃ¶ngÃ¼
# Her turda bir satÄ±r okunur (tÃ¼m dosya belleÄŸe alÄ±nmaz!)
print("Dosya iÃ§eriÄŸi:")
for satir in dosya:
    print(satir.strip())  # Her satÄ±rÄ± iÅŸle

# DosyayÄ± kapat
dosya.close()
````

# **7.4 with Ä°fadesi (Context Manager)**

`with` ifadesi, dosyayÄ± **otomatik olarak kapatÄ±r**. Hata oluÅŸsa bile dosya kapatÄ±lÄ±r.

## AvantajlarÄ±

1. `close()` yazmaya gerek yok
2. Hata durumunda bile dosya kapanÄ±r
3. Daha temiz ve gÃ¼venli kod

```python
with open("dosya.txt", "r") as dosya:
    icerik = dosya.read()
# Blok dÄ±ÅŸÄ±na Ã§Ä±kÄ±nca dosya otomatik kapanÄ±r
```

````python
# with ifadesi - Ã–nerilen yÃ¶ntem
# as anahtar kelimesi ile dosya deÄŸiÅŸkene atanÄ±r

# Yazma Ã¶rneÄŸi
with open("notlar.txt", "w") as dosya:
    dosya.write("Python Ã¶ÄŸreniyorum\n")
    dosya.write("with ifadesi Ã§ok kullanÄ±ÅŸlÄ±\n")
    dosya.write("Dosya otomatik kapanÄ±r\n")
# Blok dÄ±ÅŸÄ±na Ã§Ä±kÄ±nca dosya otomatik kapandÄ±!

print("Dosya yazÄ±ldÄ± ve otomatik kapandÄ±.")

# Okuma Ã¶rneÄŸi
with open("notlar.txt", "r") as dosya:
    icerik = dosya.read()
# Blok dÄ±ÅŸÄ±na Ã§Ä±kÄ±nca dosya otomatik kapandÄ±!

print("\n=== Ä°Ã§erik ===")
print(icerik)
````

````python
# with ile birden fazla dosya aÃ§ma
# Ä°Ã§ iÃ§e with veya tek satÄ±rda birden fazla dosya

# Kaynak dosya oluÅŸtur
with open("kaynak.txt", "w") as dosya:
    dosya.write("Bu kaynak dosyadaki veri.\n")
    dosya.write("Hedef dosyaya kopyalanacak.\n")

# Bir dosyadan okuyup diÄŸerine yaz (kopyalama)
with open("kaynak.txt", "r") as kaynak:
    with open("hedef.txt", "w") as hedef:
        for satir in kaynak:
            hedef.write(satir)

# KontrolÃ¼ ile hedef dosyayÄ± oku
with open("hedef.txt", "r") as dosya:
    print("Hedef dosya iÃ§eriÄŸi:")
    print(dosya.read())
````

# **7.5 Hata YÃ¶netimi**

Dosya iÅŸlemlerinde oluÅŸabilecek hatalar:

| Hata | AÃ§Ä±klama |
|------|----------|
| `FileNotFoundError` | Dosya bulunamadÄ± |
| `PermissionError` | Ä°zin hatasÄ± |
| `IOError` | GiriÅŸ/Ã‡Ä±kÄ±ÅŸ hatasÄ± |

`try-except` bloÄŸu ile hatalar yakalanÄ±r.

````python
# FileNotFoundError - Dosya bulunamadÄ± hatasÄ±
# try-except ile yakalama

dosya_adi = "olmayan_dosya.txt"

try:
    # Olmayan dosyayÄ± okumaya Ã§alÄ±ÅŸ
    with open(dosya_adi, "r") as dosya:
        icerik = dosya.read()
        print(icerik)
except FileNotFoundError:
    # Hata durumunda Ã§alÄ±ÅŸacak kod
    print(f"Hata: '{dosya_adi}' adlÄ± dosya bulunamadÄ±!")
    print("LÃ¼tfen dosya adÄ±nÄ± kontrol edin.")
````

````python
# Genel hata yakalama ile dosya iÅŸlemi
# finally bloÄŸu her durumda Ã§alÄ±ÅŸÄ±r

def dosya_oku(dosya_adi):
    """DosyayÄ± gÃ¼venli ÅŸekilde okur."""
    try:
        with open(dosya_adi, "r") as dosya:
            return dosya.read()
    except FileNotFoundError:
        return f"Hata: '{dosya_adi}' bulunamadÄ±!"
    except PermissionError:
        return f"Hata: '{dosya_adi}' iÃ§in izin yok!"
    except Exception as e:
        return f"Beklenmeyen hata: {e}"

# Test
print(dosya_oku("notlar.txt"))  # Var olan dosya
print()
print(dosya_oku("yok.txt"))     # Olmayan dosya
````

# **7.6 Dosya Metot ve Nitelikleri**

Dosya nesnelerinin faydalÄ± metot ve nitelikleri:

| Metot/Nitelik | AÃ§Ä±klama |
|---------------|----------|
| `read()` | TÃ¼mÃ¼nÃ¼ oku |
| `readline()` | Tek satÄ±r oku |
| `readlines()` | Liste olarak oku |
| `write()` | Yaz |
| `writelines()` | Liste yaz |
| `seek(n)` | Ä°mleci n. bayta taÅŸÄ± |
| `tell()` | Ä°mleÃ§ konumunu ver |
| `closed` | Dosya kapalÄ± mÄ±? |
| `name` | Dosya adÄ± |

````python
# Dosya nitelikleri ve seek/tell
# Ä°mleÃ§ konumu yÃ¶netimi

with open("ornek.txt", "r") as dosya:
    # Dosya adÄ±
    print(f"Dosya adÄ±: {dosya.name}")
    
    # Dosya modu
    print(f"Dosya modu: {dosya.mode}")
    
    # Ä°mleÃ§ konumu (baÅŸlangÄ±Ã§ta 0)
    print(f"Ä°mleÃ§ konumu: {dosya.tell()}")
    
    # 10 karakter oku
    print(f"\nÄ°lk 10 karakter: '{dosya.read(10)}'")
    print(f"Ä°mleÃ§ konumu: {dosya.tell()}")  # 10'a ilerledi
    
    # Ä°mleci baÅŸa al
    dosya.seek(0)  # 0. bayta git
    print(f"\nseek(0) sonrasÄ±: {dosya.tell()}")
    
    # Tekrar oku
    print(f"BaÅŸtan 5 karakter: '{dosya.read(5)}'")
````

````python
# writelines() metodu - Liste yazma
# Liste Ã¶ÄŸelerini dosyaya yazar (\n eklemez!)

satirlar = [
    "Birinci satÄ±r\n",   # \n elle eklenmeli!
    "Ä°kinci satÄ±r\n",
    "ÃœÃ§Ã¼ncÃ¼ satÄ±r\n"
]

# writelines() ile yaz
with open("liste_dosya.txt", "w") as dosya:
    dosya.writelines(satirlar)  # Listeyi yaz

# Kontrol
with open("liste_dosya.txt", "r") as dosya:
    print("Dosya iÃ§eriÄŸi:")
    print(dosya.read())
````

# **7.7 Pratik Ã–rnekler**

````python
# Pratik Ã–rnek 1: GÃ¼nlÃ¼k (Log) Sistemi
# Tarih-saat damgalÄ± kayÄ±t tutma

from datetime import datetime

def log_yaz(mesaj, dosya_adi="log.txt"):
    """Log dosyasÄ±na tarih-saat damgalÄ± mesaj yazar."""
    # Åu anki zaman
    zaman = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
    
    # Log formatÄ±: [ZAMAN] MESAJ
    log_satiri = f"[{zaman}] {mesaj}\n"
    
    # Dosyaya ekle (append modu)
    with open(dosya_adi, "a") as dosya:
        dosya.write(log_satiri)
    
    print(f"Log yazÄ±ldÄ±: {mesaj}")

def log_oku(dosya_adi="log.txt"):
    """Log dosyasÄ±nÄ± okur ve gÃ¶sterir."""
    try:
        with open(dosya_adi, "r") as dosya:
            print("=== Log KayÄ±tlarÄ± ===")
            print(dosya.read())
    except FileNotFoundError:
        print("HenÃ¼z log kaydÄ± yok!")

# Test
log_yaz("Program baÅŸlatÄ±ldÄ±")
log_yaz("KullanÄ±cÄ± giriÅŸ yaptÄ±")
log_yaz("Dosya iÅŸlemi tamamlandÄ±")
print()
log_oku()
````

````python
# Pratik Ã–rnek 2: Basit Not Defteri
# Ekleme, listeleme, silme iÅŸlemleri

NOTLAR_DOSYA = "notlarim.txt"

def not_ekle(not_metni):
    """Yeni not ekler."""
    with open(NOTLAR_DOSYA, "a") as dosya:
        dosya.write(not_metni + "\n")
    print(f"Not eklendi: {not_metni}")

def notlari_listele():
    """TÃ¼m notlarÄ± listeler."""
    try:
        with open(NOTLAR_DOSYA, "r") as dosya:
            satirlar = dosya.readlines()
            if not satirlar:
                print("HenÃ¼z not yok!")
                return
            print("=== NotlarÄ±nÄ±z ===")
            for i, satir in enumerate(satirlar, 1):
                print(f"{i}. {satir.strip()}")
    except FileNotFoundError:
        print("HenÃ¼z not yok!")

def notlari_temizle():
    """TÃ¼m notlarÄ± siler."""
    with open(NOTLAR_DOSYA, "w") as dosya:
        pass  # BoÅŸ dosya oluÅŸtur
    print("TÃ¼m notlar silindi!")

# Test
not_ekle("Python Ã¶ÄŸren")
not_ekle("Dosya iÅŸlemlerini pratik yap")
not_ekle("Proje oluÅŸtur")
print()
notlari_listele()
````

````python
# Pratik Ã–rnek 3: Kelime SayacÄ±
# Dosyadaki kelime ve satÄ±r sayÄ±sÄ±nÄ± hesapla

def dosya_istatistik(dosya_adi):
    """Dosya istatistiklerini hesaplar."""
    try:
        with open(dosya_adi, "r") as dosya:
            icerik = dosya.read()
            
            # SatÄ±r sayÄ±sÄ±
            satirlar = icerik.split("\n")
            satir_sayisi = len([s for s in satirlar if s.strip()])
            
            # Kelime sayÄ±sÄ±
            kelimeler = icerik.split()
            kelime_sayisi = len(kelimeler)
            
            # Karakter sayÄ±sÄ±
            karakter_sayisi = len(icerik)
            
            print(f"=== {dosya_adi} Ä°statistikleri ===")
            print(f"SatÄ±r sayÄ±sÄ±: {satir_sayisi}")
            print(f"Kelime sayÄ±sÄ±: {kelime_sayisi}")
            print(f"Karakter sayÄ±sÄ±: {karakter_sayisi}")
            
    except FileNotFoundError:
        print(f"Hata: '{dosya_adi}' bulunamadÄ±!")

# Test
dosya_istatistik("ornek.txt")
````

````python
# ============================================================================
# ğŸŸ¨ CEVAP ANAHTARI â€“ Dosya Ä°ÅŸlemleri
# ============================================================================
# YukarÄ±daki gÃ¶revlerin tamamlanmÄ±ÅŸ Ã§Ã¶zÃ¼mleri aÅŸaÄŸÄ±da verilmiÅŸtir.
# ============================================================================




# ============================================================================
# ğŸŸ¨ CEVAP 1: Dosya oluÅŸturma ve yazma â€“ "w" kipi
# ============================================================================
dosya = open("gorev.txt", "w")
dosya.write("Birinci satÄ±r\n")
dosya.write("Ä°kinci satÄ±r\n")
dosya.write("ÃœÃ§Ã¼ncÃ¼ satÄ±r\n")
dosya.close()
print("Dosya oluÅŸturuldu!")

print(f"\n{'-'*82}\n")




# ============================================================================
# ğŸŸ¨ CEVAP 2: Dosya ekleme â€“ "a" kipi
# ============================================================================
dosya = open("gorev.txt", "a")
dosya.write("DÃ¶rdÃ¼ncÃ¼ satÄ±r\n")
dosya.write("BeÅŸinci satÄ±r\n")
dosya.close()
print("SatÄ±rlar eklendi!")

print(f"\n{'-'*82}\n")




# ============================================================================
# ğŸŸ¨ CEVAP 3: read() metodu â€“ TÃ¼m dosyayÄ± okuma
# ============================================================================
dosya = open("gorev.txt", "r")
icerik = dosya.read()
dosya.close()
print(icerik)

print(f"\n{'-'*82}\n")




# ============================================================================
# ğŸŸ¨ CEVAP 4: readline() metodu â€“ SatÄ±r satÄ±r okuma
# ============================================================================
dosya = open("gorev.txt", "r")
satir1 = dosya.readline()
satir2 = dosya.readline()
dosya.close()
print("1. satÄ±r:", satir1.strip())
print("2. satÄ±r:", satir2.strip())

print(f"\n{'-'*82}\n")




# ============================================================================
# ğŸŸ¨ CEVAP 5: readlines() metodu â€“ Liste olarak okuma
# ============================================================================
dosya = open("gorev.txt", "r")
satirlar = dosya.readlines()
dosya.close()
print(f"SatÄ±r sayÄ±sÄ±: {len(satirlar)}")
for i, satir in enumerate(satirlar, 1):
    print(f"{i}. {satir.strip()}")

print(f"\n{'-'*82}\n")




# ============================================================================
# ğŸŸ¨ CEVAP 6: Dosya Ã¼zerinde doÄŸrudan dÃ¶ngÃ¼
# ============================================================================
dosya = open("gorev.txt", "r")
for satir in dosya:
    print(satir.strip())
dosya.close()

print(f"\n{'-'*82}\n")




# ============================================================================
# ğŸŸ¨ CEVAP 7: with ifadesi â€“ Yazma
# ============================================================================
with open("with_ornek.txt", "w") as dosya:
    dosya.write("Python Ã¶ÄŸreniyorum\n")
    dosya.write("Dosya iÅŸlemleri Ã§ok kolay\n")
    dosya.write("with ifadesi gÃ¼venli\n")
print("Dosya yazÄ±ldÄ± ve otomatik kapandÄ±.")

print(f"\n{'-'*82}\n")




# ============================================================================
# ğŸŸ¨ CEVAP 8: with ifadesi â€“ Okuma
# ============================================================================
with open("with_ornek.txt", "r") as dosya:
    icerik = dosya.read()
print(icerik)

print(f"\n{'-'*82}\n")




# ============================================================================
# ğŸŸ¨ CEVAP 9: with ile dosya kopyalama
# ============================================================================
with open("gorev.txt", "r") as kaynak:
    with open("gorev_kopya.txt", "w") as hedef:
        for satir in kaynak:
            hedef.write(satir)
print("Dosya kopyalandÄ±!")

print(f"\n{'-'*82}\n")




# ============================================================================
# ğŸŸ¨ CEVAP 10: try-except â€“ FileNotFoundError
# ============================================================================
try:
    with open("olmayan.txt", "r") as dosya:
        print(dosya.read())
except FileNotFoundError:
    print("Hata: Dosya bulunamadÄ±!")

print(f"\n{'-'*82}\n")




# ============================================================================
# ğŸŸ¨ CEVAP 11: seek() ve tell()
# ============================================================================
with open("gorev.txt", "r") as dosya:
    print(f"Ä°lk 10: '{dosya.read(10)}'")
    print(f"Ä°mleÃ§: {dosya.tell()}")
    dosya.seek(0)
    print(f"BaÅŸtan 5: '{dosya.read(5)}'")

print(f"\n{'-'*82}\n")




# ============================================================================
# ğŸŸ¨ CEVAP 12: writelines() metodu
# ============================================================================
satirlar = ["SatÄ±r A\n", "SatÄ±r B\n", "SatÄ±r C\n"]
with open("liste_ornek.txt", "w") as dosya:
    dosya.writelines(satirlar)
with open("liste_ornek.txt", "r") as dosya:
    print(dosya.read())

print(f"\n{'-'*82}\n")




# ============================================================================
# ğŸŸ¨ CEVAP 13: Pratik Ã–rnek â€“ Kelime sayacÄ±
# ============================================================================
with open("gorev.txt", "r") as dosya:
    icerik = dosya.read()
satirlar = [s for s in icerik.split("\n") if s.strip()]
kelimeler = icerik.split()
print(f"SatÄ±r: {len(satirlar)}")
print(f"Kelime: {len(kelimeler)}")
print(f"Karakter: {len(icerik)}")
````
